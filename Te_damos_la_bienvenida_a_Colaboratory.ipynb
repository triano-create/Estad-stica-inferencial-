## Pruebas de hipótesis - explicación y código

### Subtask:
Explicar los pasos 1-4 de las pruebas de hipótesis (definición de hipótesis, nivel de significancia, estadístico de prueba, valor crítico y región de rechazo). Generar código de ejemplo usando `scipy.stats` y `matplotlib` para ilustrar estos pasos.


**Reasoning**:
Explain steps 1-4 of hypothesis testing and generate code to illustrate these steps with a hypothetical scenario, including plotting the distribution and the rejection region.



```
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Explain Hypothesis Testing Steps 1-4
print("Pasos 1-4 de las Pruebas de Hipótesis:\n")

print("Paso 1: Definir las Hipótesis")
print("   - Hipótesis Nula (H0): Es la afirmación que se asume como verdadera hasta que haya evidencia suficiente en su contra. A menudo representa el status quo o la ausencia de efecto.")
print("   - Hipótesis Alternativa (H1): Es la afirmación que se desea probar. Representa lo que se concluye si se rechaza la hipótesis nula. Puede ser unilateral (mayor que o menor que) o bilateral (diferente de).\n")

print("Paso 2: Establecer el Nivel de Significancia (Alpha - α)")
print("   - Alpha es la probabilidad máxima de cometer un error de Tipo I (rechazar la hipótesis nula cuando es verdadera). Los niveles comunes son 0.05 (5%), 0.01 (1%) o 0.10 (10%). Un valor de alpha más bajo requiere evidencia más fuerte para rechazar H0.\n")

print("Paso 3: Identificar el Estadístico de Prueba")
print("   - El estadístico de prueba es un valor calculado a partir de los datos de la muestra que se utiliza para evaluar cuánto se desvía la muestra de lo esperado bajo la hipótesis nula. La elección del estadístico de prueba depende del tipo de datos y de la pregunta de investigación (por ejemplo, estadístico z, estadístico t, estadístico chi-cuadrado, etc.).\n")

print("Paso 4: Determinar el Valor Crítico y la Región de Rechazo")
print("   - El valor crítico es un umbral en la distribución del estadístico de prueba que determina si se rechaza la hipótesis nula. Está basado en el nivel de significancia y la distribución del estadístico de prueba (por ejemplo, distribución normal estándar o distribución t de Student).")
print("   - La región de rechazo es el conjunto de valores del estadístico de prueba para los cuales se rechaza la hipótesis nula. Si el estadístico de prueba cae en la región de rechazo, se rechaza H0.\n")


# Generate example code to illustrate these concepts
print("Ejemplo de Código para Ilustrar Conceptos:\n")

# Hypothetical scenario: Testing if the mean of a population is different from a known value (e.g., 50)
# We'll assume we know the population standard deviation for simplicity, so we use a z-test.

# Step 1: Define Hypotheses
# H0: μ = 50 (The population mean is 50)
# H1: μ ≠ 50 (The population mean is not 50) - This is a two-tailed test

# Step 2: Set Significance Level (Alpha)
alpha = 0.05
print(f"Nivel de significancia (alpha): {alpha}")

# Step 3 & 4: Determine Critical Value and Rejection Region (for a two-tailed test)
# For a two-tailed z-test, the critical values are such that alpha/2 is in each tail.
critical_value_lower = norm.ppf(alpha / 2)
critical_value_upper = norm.ppf(1 - alpha / 2)

print(f"Valor crítico inferior (Z): {critical_value_lower:.4f}")
print(f"Valor crítico superior (Z): {critical_value_upper:.4f}")
print(f"Región de rechazo: Z < {critical_value_lower:.4f} o Z > {critical_value_upper:.4f}\n")


# Plot the standard normal distribution and shade the rejection region
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1) # Standard normal distribution (mean=0, std dev=1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Normal Estándar (Z)')

# Shade the rejection region
# Shade the lower tail
x_reject_lower = x[x < critical_value_lower]
y_reject_lower = norm.pdf(x_reject_lower, 0, 1)
plt.fill_between(x_reject_lower, 0, y_reject_lower, color='red', alpha=0.5, label='Región de Rechazo (α/2)')

# Shade the upper tail
x_reject_upper = x[x > critical_value_upper]
y_reject_upper = norm.pdf(x_reject_upper, 0, 1)
plt.fill_between(x_reject_upper, 0, y_reject_upper, color='red', alpha=0.5) # No label needed for the second part of the same region


plt.title('Distribución Normal Estándar con Región de Rechazo (Test Bilateral)')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.axvline(critical_value_lower, color='black', linestyle='--', label=f'Valor Crítico Inferior ({critical_value_lower:.2f})')
plt.axvline(critical_value_upper, color='black', linestyle='--', label=f'Valor Crítico Superior ({critical_value_upper:.2f})')
plt.legend()
plt.grid(True)
plt.show()

print("La gráfica muestra la distribución normal estándar. Las áreas sombreadas en rojo representan la región de rechazo.")
print("Si el estadístico de prueba Z calculado a partir de los datos de la muestra cae dentro de estas áreas sombreadas,")
print(f"rechazaríamos la hipótesis nula (H0) al nivel de significancia de {alpha*100}%.")
```

## Pruebas de hipótesis - cálculo y visualización

### Subtask:
Explicar los pasos 5-6 de las pruebas de hipótesis (cálculo del estadístico de prueba y valor p, graficación). Generar código para calcular el estadístico de prueba y el valor p, y para graficar la distribución con la región de rechazo, valor crítico y estadístico calculado.


**Reasoning**:
Print the explanation for steps 5 and 6 of hypothesis testing, then generate the code to calculate the test statistic and p-value, and finally create the plot to visualize the distribution, critical values, test statistic, and rejection region as per the instructions.



```
# Explain Hypothesis Testing Steps 5-6
print("\nPasos 5-6 de las Pruebas de Hipótesis:\n")

print("Paso 5: Calcular el Estadístico de Prueba y el Valor p")
print("   - Calcular el estadístico de prueba utilizando los datos de la muestra. La fórmula específica depende del tipo de prueba que se esté realizando.")
print("   - El valor p es la probabilidad de obtener un estadístico de prueba tan extremo, o más extremo, como el observado en la muestra, asumiendo que la hipótesis nula es verdadera. Un valor p pequeño (típicamente menor que alpha) proporciona evidencia contra H0.\n")

print("Paso 6: Tomar una Decisión y Graficar los Resultados")
print("   - Comparar el estadístico de prueba calculado con el valor crítico o comparar el valor p con el nivel de significancia (alpha).")
print("   - Si el estadístico de prueba cae en la región de rechazo (o si el valor p <= alpha), se rechaza la hipótesis nula.")
print("   - Si el estadístico de prueba no cae en la región de rechazo (o si el valor p > alpha), no se rechaza la hipótesis nula.")
print("   - Graficar la distribución, el valor crítico, el estadístico de prueba y la región de rechazo es crucial para visualizar la decisión y entender dónde cae el estadístico de prueba en relación con la región de rechazo. Esto ayuda a comunicar los resultados de la prueba de hipótesis de manera efectiva.\n")


# Simulate sample data for a hypothetical scenario (Steps 2-4 were covered in the previous section)
# Assume we are performing a two-tailed z-test at alpha = 0.05
# H0: mu = 50
# H1: mu != 50
# Known population standard deviation: sigma = 10

n = 30      # Sample size
x_bar = 53  # Sample mean
mu_0 = 50   # Population mean under the null hypothesis
sigma = 10  # Population standard deviation

print(f"Escenario simulado: Tamaño de muestra n = {n}, Media de la muestra x̄ = {x_bar}")
print(f"Hipótesis Nula H0: μ = {mu_0}, Desviación estándar de la población σ = {sigma}\n")


# Step 5: Calculate the Test Statistic (Z-statistic for a sample mean with known population standard deviation)
# Formula: Z = (x_bar - mu_0) / (sigma / sqrt(n))
z_statistic = (x_bar - mu_0) / (sigma / np.sqrt(n))

# Calculate the p-value (for a two-tailed test)
# The p-value is the probability of observing a Z-statistic as extreme as the one calculated, in either tail.
# We calculate the probability of the upper tail (Z > z_statistic) using sf (survival function, which is 1 - cdf)
# and multiply by 2 for a two-tailed test.
p_value = 2 * norm.sf(abs(z_statistic))


# Step 6: Plot the distribution, critical values, test statistic, and rejection region
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1) # Standard normal distribution (mean=0, std dev=1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Normal Estándar (Z)')

# Shade the rejection region (using critical_value_lower and critical_value_upper from previous step)
x_reject_lower = x[x < critical_value_lower]
y_reject_lower = norm.pdf(x_reject_lower, 0, 1)
plt.fill_between(x_reject_lower, 0, y_reject_lower, color='red', alpha=0.5, label='Región de Rechazo (α/2)')

x_reject_upper = x[x > critical_value_upper]
y_reject_upper = norm.pdf(x_reject_upper, 0, 1)
plt.fill_between(x_reject_upper, 0, y_reject_upper, color='red', alpha=0.5)


# Add vertical lines for critical values and the test statistic
plt.axvline(critical_value_lower, color='black', linestyle='--', label=f'Valor Crítico Inferior ({critical_value_lower:.2f})')
plt.axvline(critical_value_upper, color='black', linestyle='--', label=f'Valor Crítico Superior ({critical_value_upper:.2f})')
plt.axvline(z_statistic, color='green', linestyle='-', label=f'Estadístico de Prueba Z ({z_statistic:.2f})', linewidth=2)


plt.title('Distribución Normal Estándar con Región de Rechazo, Valores Críticos y Estadístico de Prueba')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

# Print the calculated test statistic and p-value
print(f"Estadístico de Prueba Z calculado: {z_statistic:.4f}")
print(f"Valor p calculado (bilateral): {p_value:.4f}\n")

# Interpretation
print("Interpretación:")
print(f"- El estadístico de prueba Z calculado es {z_statistic:.4f}.")
print(f"- Los valores críticos para un nivel de significancia alpha = {alpha} son aproximadamente {critical_value_lower:.4f} y {critical_value_upper:.4f}.")
print(f"- El valor p calculado es {p_value:.4f}.")

if abs(z_statistic) > critical_value_upper: # Check if |Z_statistic| > |critical_value| for two-tailed test
    print(f"- Como el valor absoluto del estadístico de prueba (|{abs(z_statistic):.4f}|) es mayor que el valor crítico superior ({critical_value_upper:.4f}), o equivalentemente, como el valor p ({p_value:.4f}) es menor o igual que alpha ({alpha}), rechazamos la hipótesis nula (H0).")
    print("  Hay suficiente evidencia para concluir que la media de la población es significativamente diferente de 50.")
else:
    print(f"- Como el estadístico de prueba ({z_statistic:.4f}) cae entre los valores críticos ({critical_value_lower:.4f} y {critical_value_upper:.4f}), o equivalentemente, como el valor p ({p_value:.4f}) es mayor que alpha ({alpha}), no rechazamos la hipótesis nula (H0).")
    print("  No hay suficiente evidencia para concluir que la media de la población es significativamente diferente de 50.")
```

## Intervalos de confianza

### Subtask:
Explicar el concepto de intervalos de confianza. Generar código para calcular intervalos de confianza usando `scipy.stats`. Explicar el nivel de confianza y los parámetros utilizados.


**Reasoning**:
Explain the concept of confidence intervals and calculate one using the previously simulated data and scipy.stats, then explain the parameters and interpret the result, fulfilling all instructions for this subtask.



```
from scipy.stats import norm

# 1. Explain Confidence Intervals
print("Intervalos de Confianza\n")
print("Un intervalo de confianza es un rango de valores que probablemente contenga un parámetro de población desconocido (como la media o la proporción) con un cierto nivel de confianza.")
print("Proporciona una estimación de la precisión de la estimación puntual obtenida de una muestra.")
print("El propósito es cuantificar la incertidumbre asociada con una estimación de la muestra. En lugar de dar un solo número (la estimación puntual), da un rango de valores plausibles para el parámetro de la población.")
print("\nEl Nivel de Confianza:")
print("   - El nivel de confianza (comúnmente 90%, 95% o 99%) es la probabilidad de que, si se repitiera el proceso de muestreo y construcción del intervalo un gran número de veces, el intervalo resultante contenga el verdadero parámetro de la población.")
print("   - Por ejemplo, un intervalo de confianza del 95% significa que si tomáramos 100 muestras aleatorias de la misma población y calculáramos un intervalo de confianza para cada muestra, esperaríamos que aproximadamente 95 de esos intervalos contuvieran el verdadero parámetro de la población.")
print("   - No es la probabilidad de que el verdadero parámetro de la población caiga dentro de un intervalo específico calculado. El parámetro de la población es un valor fijo (aunque desconocido), mientras que el intervalo de confianza varía de una muestra a otra.")
print("   - Un nivel de confianza más alto resulta en un intervalo más amplio (mayor incertidumbre), mientras que un nivel de confianza más bajo resulta en un intervalo más estrecho (menor incertidumbre), asumiendo que todo lo demás es constante.\n")

# 2. Calculate a Confidence Interval for the Population Mean
# Using the simulated data: x_bar, sigma, n from the previous step
# Let's use a 95% confidence level, which corresponds to alpha = 0.05

confidence_level = 0.95
alpha_ci = 1 - confidence_level
print(f"Calculando el intervalo de confianza para la media poblacional (μ) con:")
print(f"  Media muestral (x̄): {x_bar}")
print(f"  Desviación estándar poblacional (σ): {sigma}")
print(f"  Tamaño de muestra (n): {n}")
print(f"  Nivel de confianza: {confidence_level*100}%\n")


# For a known population standard deviation (sigma) and a large sample size (n >= 30),
# we can use the Z-distribution to calculate the confidence interval.
# The formula for a confidence interval for the mean is: x_bar ± Z * (sigma / sqrt(n))
# Where Z is the critical value from the standard normal distribution corresponding to the desired confidence level.

# We need the Z-value such that (1 - alpha_ci) of the area is between -Z and +Z.
# This means alpha_ci/2 is in each tail. The value we need is at the (1 - alpha_ci/2) percentile.
z_critical_ci = norm.ppf(1 - alpha_ci / 2)

# Calculate the margin of error
margin_of_error = z_critical_ci * (sigma / np.sqrt(n))

# Calculate the confidence interval bounds
lower_bound_ci = x_bar - margin_of_error
upper_bound_ci = x_bar + margin_of_error

# 3. Print the Calculated Confidence Interval
print(f"El valor crítico Z para un nivel de confianza del {confidence_level*100}% es aproximadamente {z_critical_ci:.4f}")
print(f"El margen de error es: {margin_of_error:.4f}")
print(f"El intervalo de confianza del {confidence_level*100}% para la media poblacional es: [{lower_bound_ci:.4f}, {upper_bound_ci:.4f}]\n")


# Alternatively, use scipy.stats.norm.interval
# This function directly calculates the confidence interval for the mean of a normal distribution
# or for a sample mean when sigma is known.
# Parameters: confidence level, loc (mean), scale (standard error = sigma / sqrt(n))
confidence_interval_scipy = norm.interval(confidence_level, loc=x_bar, scale=sigma / np.sqrt(n))

print(f"Intervalo de confianza calculado usando scipy.stats.norm.interval: {confidence_interval_scipy}\n")


# 4. Explain the Parameters Used
print("Explicación de los Parámetros Utilizados:")
print(f"  - Media muestral (x̄ = {x_bar}): Es la media de los datos observados en nuestra muestra.")
print(f"  - Desviación estándar poblacional (σ = {sigma}): Es la desviación estándar de la población completa. En escenarios reales, a menudo es desconocida y se usa la desviación estándar muestral (s) en su lugar, lo que requiere usar la distribución t de Student.")
print(f"  - Tamaño de muestra (n = {n}): Es el número de observaciones en nuestra muestra. Un tamaño de muestra mayor generalmente resulta en un intervalo de confianza más estrecho (más preciso).")
print(f"  - Nivel de confianza ({confidence_level*100}%): Es el grado de seguridad que queremos tener de que el intervalo contenga el verdadero parámetro de la población. Un nivel más alto requiere un intervalo más amplio.\n")
print("Cómo afectan estos parámetros al intervalo:")
print("  - x̄ determina el centro del intervalo.")
print("  - σ y n afectan el ancho del intervalo a través del error estándar (σ/√n). Un σ más grande o un n más pequeño aumentan el error estándar y hacen el intervalo más amplio.")
print("  - El nivel de confianza afecta el valor crítico Z. Un nivel de confianza más alto requiere un Z más grande, lo que aumenta el margen de error y hace el intervalo más amplio.\n")


# 5. Interpret the Calculated Confidence Interval
print("Interpretación del Intervalo de Confianza:")
print(f"Basado en nuestra muestra, tenemos un {confidence_level*100}% de confianza en que el verdadero valor de la media poblacional (μ) se encuentra entre {lower_bound_ci:.4f} y {upper_bound_ci:.4f}.")
print("Esto significa que si repitiéramos este estudio muchas veces, el 95% de los intervalos de confianza construidos de esta manera contendrían la verdadera media poblacional.")
print("No podemos decir que hay un 95% de probabilidad de que la verdadera media poblacional caiga dentro de *este* intervalo específico, ya que la verdadera media es un valor fijo y o bien está en el intervalo o no lo está.")
print("El intervalo nos da un rango de valores plausibles para la media poblacional, dada la evidencia de nuestra muestra.")
```

Explicar el concepto de tamaño de muestra. Generar código para calcular el tamaño de muestra para diferentes escenarios. Explicar las fórmulas y parámetros utilizados.

```
from scipy.stats import norm
import math

# 3. Calculate sample size for estimating a population mean
print("\nCálculo del Tamaño de Muestra para Estimar una Media Poblacional\n")

# Formula for sample size for estimating mean:
# n = (Z * σ / E)^2
# Where:
# n = required sample size
# Z = Z-score corresponding to the desired confidence level
# σ = population standard deviation (or an estimate)
# E = desired margin of error

print("Fórmula utilizada: n = (Z * σ / E)^2")
print("Donde:")
print("  n = Tamaño de muestra requerido")
print("  Z = Puntuación Z para el nivel de confianza deseado (por ejemplo, 1.96 para 95% de confianza)")
print("  σ = Desviación estándar de la población (o una estimación previa)")
print("  E = Margen de error máximo aceptable\n")


# Example Calculation 1: Estimate mean with 95% confidence, margin of error 5, std dev 10
desired_confidence_level = 0.95
desired_margin_of_error = 5
estimated_sigma = 10 # We need an estimate of the population standard deviation

# Find the Z-score for the desired confidence level
# For a two-tailed interval, the alpha is 1 - confidence_level
alpha_ss = 1 - desired_confidence_level
z_score_estimate = norm.ppf(1 - alpha_ss / 2)

# Calculate the required sample size
n_estimate_mean = math.ceil((z_score_estimate * estimated_sigma / desired_margin_of_error) ** 2) # Use math.ceil to round up to the nearest whole number

print(f"Ejemplo 1: Estimar la media poblacional")
print(f"  Nivel de confianza deseado: {desired_confidence_level*100}%")
print(f"  Margen de error deseado (E): {desired_margin_of_error}")
print(f"  Desviación estándar estimada (σ): {estimated_sigma}")
print(f"  Puntuación Z para {desired_confidence_level*100}%: {z_score_estimate:.4f}")
print(f"  Tamaño de muestra requerido: ({z_score_estimate:.4f} * {estimated_sigma} / {desired_margin_of_error})^2 = {n_estimate_mean}\n")


# Example Calculation 2: Different parameters
desired_confidence_level_ex2 = 0.99
desired_margin_of_error_ex2 = 2
estimated_sigma_ex2 = 15

alpha_ss_ex2 = 1 - desired_confidence_level_ex2
z_score_estimate_ex2 = norm.ppf(1 - alpha_ss_ex2 / 2)
n_estimate_mean_ex2 = math.ceil((z_score_estimate_ex2 * estimated_sigma_ex2 / desired_margin_of_error_ex2) ** 2)

print(f"Ejemplo 2: Estimar la media poblacional con parámetros diferentes")
print(f"  Nivel de confianza deseado: {desired_confidence_level_ex2*100}%")
print(f"  Margen de error deseado (E): {desired_margin_of_error_ex2}")
print(f"  Desviación estándar estimada (σ): {estimated_sigma_ex2}")
print(f"  Puntuación Z para {desired_confidence_level_ex2*100}%: {z_score_estimate_ex2:.4f}")
print(f"  Tamaño de muestra requerido: ({z_score_estimate_ex2:.4f} * {estimated_sigma_ex2} / {desired_margin_of_error_ex2})^2 = {n_estimate_mean_ex2}\n")

print("Observaciones:")
print("- Un margen de error más pequeño (Ejemplo 2 vs Ejemplo 1) requiere un tamaño de muestra significativamente mayor.")
print("- Un nivel de confianza más alto (Ejemplo 2 vs Ejemplo 1) requiere un tamaño de muestra mayor.")
print("- Una mayor variabilidad (σ) en la población también requiere un tamaño de muestra mayor (comparando si σ fuera diferente en los ejemplos con otros parámetros iguales).")
```

```
# 4. Calculate sample size for a hypothesis test (one-sample z-test for mean)
# 5. Explain the formulas and parameters used in both scenarios (part 2)
# 6. Provide more examples (combined with calculation steps)

print("Cálculo del Tamaño de Muestra para una Prueba de Hipótesis (Test Z Unilateral para la Media)\n")

# We will calculate sample size for a one-sample z-test for a mean.
# This calculation requires specifying:
# - Significance level (alpha)
# - Desired power (1 - beta)
# - Effect size (delta / sigma), where delta is the minimum difference from the null hypothesis mean we want to detect.

# Formula for sample size (one-sided test, known sigma):
# n = ((Z_alpha + Z_beta) * σ / δ)^2
# Where:
# n = required sample size
# Z_alpha = Z-score corresponding to the significance level (alpha)
# Z_beta = Z-score corresponding to the desired power (1 - beta)
# σ = population standard deviation (or an estimate)
# δ = minimum detectable difference (effect size in original units)

print("Fórmula utilizada (Test Unilateral): n = ((Z_α + Z_β) * σ / δ)^2")
print("Donde:")
print("  n = Tamaño de muestra requerido")
print("  Z_α = Puntuación Z para el nivel de significancia (α)")
print("  Z_β = Puntuación Z para el poder deseado (1 - β). Beta (β) es la probabilidad de un error Tipo II.")
print("  σ = Desviación estándar de la población (o una estimación previa)")
print("  δ = Diferencia mínima que se desea detectar (Tamaño del efecto en unidades originales)\n")

# Note: For a two-sided test, the formula is slightly different, often using Z_(alpha/2).
# The formula can also be expressed in terms of standardized effect size (Cohen's d = δ / σ):
# n = ((Z_alpha + Z_beta) / d)^2

print("Alternativamente, usando el tamaño del efecto estandarizado (d = δ / σ): n = ((Z_α + Z_β) / d)^2\n")


# Example Calculation: One-sample z-test
# We want to detect a difference of 3 units from the null mean (e.g., H0: mu=50, we want to detect if mu=53 or more)
# Significance level (alpha): 0.05 (one-sided test)
# Desired power: 0.80 (beta = 0.20)
# Estimated population standard deviation (sigma): 10
# Minimum detectable difference (delta): 3

significance_level = 0.05
desired_power = 0.80
estimated_sigma = 10
minimum_detectable_difference = 3 # delta

# Find the Z-score for alpha (for a one-sided test)
z_alpha = norm.ppf(1 - significance_level)

# Find the Z-score for beta (for calculating power)
beta = 1 - desired_power
z_beta = norm.ppf(1 - beta) # This is the Z-score corresponding to the desired power

# Calculate the required sample size
n_hypothesis_test = math.ceil(((z_alpha + z_beta) * estimated_sigma / minimum_detectable_difference) ** 2)

print(f"Ejemplo: Prueba de Hipótesis (Test Z Unilateral)")
print(f"  Nivel de significancia (α): {significance_level}")
print(f"  Poder deseado (1 - β): {desired_power}")
print(f"  Desviación estándar estimada (σ): {estimated_sigma}")
print(f"  Diferencia mínima a detectar (δ): {minimum_detectable_difference}")
print(f"  Puntuación Z para α ({significance_level}): {z_alpha:.4f}")
print(f"  Puntuación Z para 1-β ({desired_power}): {z_beta:.4f}")
print(f"  Tamaño de muestra requerido: (({z_alpha:.4f} + {z_beta:.4f}) * {estimated_sigma} / {minimum_detectable_difference})^2 = {n_hypothesis_test}\n")


# Another Example: Different parameters
# Significance level (alpha): 0.01 (one-sided)
# Desired power: 0.95
# Estimated population standard deviation (sigma): 15
# Minimum detectable difference (delta): 2

significance_level_ex2 = 0.01
desired_power_ex2 = 0.95
estimated_sigma_ex2 = 15
minimum_detectable_difference_ex2 = 2

z_alpha_ex2 = norm.ppf(1 - significance_level_ex2)
beta_ex2 = 1 - desired_power_ex2
z_beta_ex2 = norm.ppf(1 - beta_ex2)
n_hypothesis_test_ex2 = math.ceil(((z_alpha_ex2 + z_beta_ex2) * estimated_sigma_ex2 / minimum_detectable_difference_ex2) ** 2)


print(f"Otro Ejemplo: Prueba de Hipótesis (Test Z Unilateral) con parámetros diferentes")
print(f"  Nivel de significancia (α): {significance_level_ex2}")
print(f"  Poder deseado (1 - β): {desired_power_ex2}")
print(f"  Desviación estándar estimada (σ): {estimated_sigma_ex2}")
print(f"  Diferencia mínima a detectar (δ): {minimum_detectable_difference_ex2}")
print(f"  Puntuación Z para α ({significance_level_ex2}): {z_alpha_ex2:.4f}")
print(f"  Puntuación Z para 1-β ({desired_power_ex2}): {z_beta_ex2:.4f}")
print(f"  Tamaño de muestra requerido: (({z_alpha_ex2:.4f} + {z_beta_ex2:.4f}) * {estimated_sigma_ex2} / {minimum_detectable_difference_ex2})^2 = {n_hypothesis_test_ex2}\n")

print("Observaciones:")
print("- Un poder deseado más alto (Ejemplo 2 vs Ejemplo 1) requiere un tamaño de muestra mayor.")
print("- Un nivel de significancia más estricto (alpha más pequeño, Ejemplo 2 vs Ejemplo 1) requiere un tamaño de muestra mayor.")
print("- La detección de una diferencia más pequeña (delta más pequeño, Ejemplo 2 vs Ejemplo 1) requiere un tamaño de muestra significativamente mayor.")
print("- Una mayor variabilidad (σ) también requiere un tamaño de muestra mayor.")


# Using statsmodels for power analysis (more comprehensive, supports various tests)
# Note: statsmodels might need installation if not already present in the environment.
# %pip install statsmodels

try:
    from statsmodels.stats.power import NormalIndPower
    from statsmodels.stats.proportion import proportion_effectsize

    print("Cálculo del Tamaño de Muestra usando statsmodels (Alternativa)\n")

    # Example using statsmodels for one-sample z-test for mean
    # We need to specify the effect size (Cohen's d = delta / sigma)
    # For the first example: delta=3, sigma=10, so Cohen's d = 3/10 = 0.3 (medium effect size)

    effect_size_statmodels = minimum_detectable_difference / estimated_sigma
    alpha_statmodels = significance_level
    power_statmodels = desired_power
    ratio = 0 # ratio of sample sizes, set to 0 for one sample test

    # Create a power analysis object for normal distribution (z-test)
    power_analysis = NormalIndPower()

    # Calculate sample size (per group, but for one sample test, it's the total sample size)
    # The 'solve_for' parameter is set to 'nobs1' to calculate the sample size for the first group (our single sample)
    # The 'alternative' parameter specifies the type of test ('two-sided', 'one-sided')
    # Note: statsmodels' NormalIndPower is primarily for two-sample tests, but setting ratio=0 allows it for one-sample.
    # For one-sample, the calculated 'nobs1' is the total sample size.

    n_statmodels = power_analysis.solve_power(
        effect_size=effect_size_statmodels,
        alpha=alpha_statmodels,
        power=power_statmodels,
        ratio=ratio, # Use ratio=0 for one-sample test
        alternative='one-sided' # or 'two-sided' depending on the test
    )

    print(f"Ejemplo usando statsmodels (Test Z Unilateral para la Media):")
    print(f"  Tamaño del efecto (Cohen's d = δ/σ): {effect_size_statmodels:.4f}")
    print(f"  Nivel de significancia (α): {alpha_statmodels}")
    print(f"  Poder deseado (1 - β): {power_statmodels}")
    print(f"  Tamaño de muestra requerido: {math.ceil(n_statmodels)}\n")


    # Example using statsmodels for the second scenario
    effect_size_statmodels_ex2 = minimum_detectable_difference_ex2 / estimated_sigma_ex2
    alpha_statmodels_ex2 = significance_level_ex2
    power_statmodels_ex2 = desired_power_ex2

    n_statmodels_ex2 = power_analysis.solve_power(
        effect_size=effect_size_statmodels_ex2,
        alpha=alpha_statmodels_ex2,
        power=power_statmodels_ex2,
        ratio=ratio,
        alternative='one-sided' # or 'two-sided'
    )
    print(f"Otro Ejemplo usando statsmodels (Test Z Unilateral) con parámetros diferentes:")
    print(f"  Tamaño del efecto (Cohen's d = δ/σ): {effect_size_statmodels_ex2:.4f}")
    print(f"  Nivel de significancia (α): {alpha_statmodels_ex2}")
    print(f"  Poder deseado (1 - β): {power_statmodels_ex2}")
    print(f"  Tamaño de muestra requerido: {math.ceil(n_statmodels_ex2)}\n")

except ImportError:
    print("La librería statsmodels no está instalada. Para usarla, ejecute: %pip install statsmodels")

print("Explicación de los Parámetros en el Cálculo del Tamaño de Muestra para Pruebas de Hipótesis:")
print("  - Nivel de Significancia (α): La probabilidad de rechazar H0 cuando es verdadera (Error Tipo I). Un α más pequeño requiere más evidencia, por lo tanto, un tamaño de muestra mayor para mantener el mismo poder.")
print("  - Poder (1 - β): La probabilidad de rechazar H0 cuando es falsa (detectar un efecto real). Un poder más alto significa una menor probabilidad de Error Tipo II (β), lo que requiere un tamaño de muestra mayor.")
print("  - Desviación Estándar (σ): La variabilidad de la población. Una mayor variabilidad hace que sea más difícil detectar una diferencia, requiriendo un tamaño de muestra mayor.")
print("  - Diferencia Mínima a Detectar (δ) o Tamaño del Efecto (d): La magnitud del efecto que el estudio está diseñado para detectar. Un efecto más pequeño requiere un tamaño de muestra mayor para ser detectado con la misma confianza y poder.")
print("  - Z_α y Z_β: Son los valores de la distribución normal estándar correspondientes a los percentiles definidos por alpha y beta, respectivamente. Se obtienen usando la función inversa de la CDF (percentile point function - ppf).")

print("\nConsideraciones adicionales para pruebas de hipótesis:")
print("- La elección de si la prueba es unilateral ('one-sided') o bilateral ('two-sided') afecta el valor Z_α y, por lo tanto, el tamaño de muestra requerido.")
print("- Si la desviación estándar de la población (σ) es desconocida, a menudo se utiliza una estimación previa de σ (quizás de estudios piloto o literatura existente).")
```

```
from scipy.stats import norm
import math
from statsmodels.stats.power import NormalIndPower

# 4. Calculate sample size for a hypothesis test (one-sample z-test for mean)
# 5. Explain the formulas and parameters used in both scenarios (part 2)
# 6. Provide more examples (combined with calculation steps)

print("Cálculo del Tamaño de Muestra para una Prueba de Hipótesis (Test Z Unilateral para la Media)\n")

# We will calculate sample size for a one-sample z-test for a mean.
# This calculation requires specifying:
# - Significance level (alpha)
# - Desired power (1 - beta)
# - Effect size (delta / sigma), where delta is the minimum difference from the null hypothesis mean we want to detect.

# Formula for sample size (one-sided test, known sigma):
# n = ((Z_alpha + Z_beta) * σ / δ)^2
# Where:
# n = required sample size
# Z_alpha = Z-score corresponding to the significance level (alpha)
# Z_beta = Z-score corresponding to the desired power (1 - beta)
# σ = population standard deviation (or an estimate)
# δ = minimum detectable difference (effect size in original units)

print("Fórmula utilizada (Test Unilateral): n = ((Z_α + Z_β) * σ / δ)^2")
print("Donde:")
print("  n = Tamaño de muestra requerido")
print("  Z_α = Puntuación Z para el nivel de significancia (α)")
print("  Z_β = Puntuación Z para el poder deseado (1 - β). Beta (β) es la probabilidad de un error Tipo II.")
print("  σ = Desviación estándar de la población (o una estimación previa)")
print("  δ = Diferencia mínima que se desea detectar (Tamaño del efecto en unidades originales)\n")

# Note: For a two-sided test, the formula is slightly different, often using Z_(alpha/2).
# The formula can also be expressed in terms of standardized effect size (Cohen's d = δ / σ):
# n = ((Z_alpha + Z_beta) / d)^2

print("Alternativamente, usando el tamaño del efecto estandarizado (d = δ / σ): n = ((Z_α + Z_β) / d)^2\n")


# Example Calculation: One-sample z-test
# We want to detect a difference of 3 units from the null mean (e.g., H0: mu=50, we want to detect if mu=53 or more)
# Significance level (alpha): 0.05 (one-sided test)
# Desired power: 0.80 (beta = 0.20)
# Estimated population standard deviation (sigma): 10
# Minimum detectable difference (delta): 3

significance_level = 0.05
desired_power = 0.80
estimated_sigma = 10
minimum_detectable_difference = 3 # delta

# Find the Z-score for alpha (for a one-sided test)
z_alpha = norm.ppf(1 - significance_level)

# Find the Z-score for beta (for calculating power)
beta = 1 - desired_power
z_beta = norm.ppf(1 - beta) # This is the Z-score corresponding to the desired power

# Calculate the required sample size
n_hypothesis_test = math.ceil(((z_alpha + z_beta) * estimated_sigma / minimum_detectable_difference) ** 2)

print(f"Ejemplo: Prueba de Hipótesis (Test Z Unilateral)")
print(f"  Nivel de significancia (α): {significance_level}")
print(f"  Poder deseado (1 - β): {desired_power}")
print(f"  Desviación estándar estimada (σ): {estimated_sigma}")
print(f"  Diferencia mínima a detectar (δ): {minimum_detectable_difference}")
print(f"  Puntuación Z para α ({significance_level}): {z_alpha:.4f}")
print(f"  Puntuación Z para 1-β ({desired_power}): {z_beta:.4f}")
print(f"  Tamaño de muestra requerido: (({z_alpha:.4f} + {z_beta:.4f}) * {estimated_sigma} / {minimum_detectable_difference})^2 = {n_hypothesis_test}\n")


# Another Example: Different parameters
# Significance level (alpha): 0.01 (one-sided)
# Desired power: 0.95
# Estimated population standard deviation (sigma): 15
# Minimum detectable difference (delta): 2

significance_level_ex2 = 0.01
desired_power_ex2 = 0.95
estimated_sigma_ex2 = 15
minimum_detectable_difference_ex2 = 2

z_alpha_ex2 = norm.ppf(1 - significance_level_ex2)
beta_ex2 = 1 - desired_power_ex2
z_beta_ex2 = norm.ppf(1 - beta_ex2)
n_hypothesis_test_ex2 = math.ceil(((z_alpha_ex2 + z_beta_ex2) * estimated_sigma_ex2 / minimum_detectable_difference_ex2) ** 2)


print(f"Otro Ejemplo: Prueba de Hipótesis (Test Z Unilateral) con parámetros diferentes")
print(f"  Nivel de significancia (α): {significance_level_ex2}")
print(f"  Poder deseado (1 - β): {desired_power_ex2}")
print(f"  Desviación estándar estimada (σ): {estimated_sigma_ex2}")
print(f"  Diferencia mínima a detectar (δ): {minimum_detectable_difference_ex2}")
print(f"  Puntuación Z para α ({significance_level_ex2}): {z_alpha_ex2:.4f}")
print(f"  Puntuación Z para 1-β ({desired_power_ex2}): {z_beta_ex2:.4f}")
print(f"  Tamaño de muestra requerido: (({z_alpha_ex2:.4f} + {z_beta_ex2:.4f}) * {estimated_sigma_ex2} / {minimum_detectable_difference_ex2})^2 = {n_hypothesis_test_ex2}\n")

print("Observaciones:")
print("- Un poder deseado más alto (Ejemplo 2 vs Ejemplo 1) requiere un tamaño de muestra mayor.")
print("- Un nivel de significancia más estricto (alpha más pequeño, Ejemplo 2 vs Ejemplo 1) requiere un tamaño de muestra mayor.")
print("- La detección de una diferencia más pequeña (delta más pequeño, Ejemplo 2 vs Ejemplo 1) requiere un tamaño de muestra significativamente mayor.")
print("- Una mayor variabilidad (σ) también requiere un tamaño de muestra mayor.")


# Using statsmodels for power analysis (more comprehensive, supports various tests)
# Note: statsmodels might need installation if not already present in the environment.
# %pip install statsmodels

try:
    from statsmodels.stats.power import NormalIndPower
    from statsmodels.stats.proportion import proportion_effectsize

    print("Cálculo del Tamaño de Muestra usando statsmodels (Alternativa)\n")

    # Example using statsmodels for one-sample z-test for mean
    # We need to specify the effect size (Cohen's d = delta / sigma)
    # For the first example: delta=3, sigma=10, so Cohen's d = 3/10 = 0.3 (medium effect size)

    effect_size_statmodels = minimum_detectable_difference / estimated_sigma
    alpha_statmodels = significance_level
    power_statmodels = desired_power
    ratio = 0 # ratio of sample sizes, set to 0 for one sample test

    # Create a power analysis object for normal distribution (z-test)
    power_analysis = NormalIndPower()

    # Calculate sample size (per group, but for one sample test, it's the total sample size)
    # The 'solve_for' parameter is set to 'nobs1' to calculate the sample size for the first group (our single sample)
    # The 'alternative' parameter specifies the type of test ('two-sided', 'larger' or 'smaller')
    # Fix: Changed 'one-sided' to 'larger' (assuming a test for a mean larger than the null)
    n_statmodels = power_analysis.solve_power(
        effect_size=effect_size_statmodels,
        alpha=alpha_statmodels,
        power=power_statmodels,
        ratio=ratio, # Use ratio=0 for one-sample test
        alternative='larger' # Valid options: 'two-sided', 'larger', 'smaller'
    )

    print(f"Ejemplo usando statsmodels (Test Z Unilateral para la Media, alternativa='larger'):")
    print(f"  Tamaño del efecto (Cohen's d = δ/σ): {effect_size_statmodels:.4f}")
    print(f"  Nivel de significancia (α): {alpha_statmodels}")
    print(f"  Poder deseado (1 - β): {power_statmodels}")
    print(f"  Tamaño de muestra requerido: {math.ceil(n_statmodels)}\n")


    # Example using statsmodels for the second scenario
    effect_size_statmodels_ex2 = minimum_detectable_difference_ex2 / estimated_sigma_ex2
    alpha_statmodels_ex2 = significance_level_ex2
    power_statmodels_ex2 = desired_power_ex2

    # Fix: Changed 'one-sided' to 'larger' (assuming a test for a mean larger than the null)
    n_statmodels_ex2 = power_analysis.solve_power(
        effect_size=effect_size_statmodels_ex2,
        alpha=alpha_statmodels_ex2,
        power=power_statmodels_ex2,
        ratio=ratio,
        alternative='larger' # Valid options: 'two-sided', 'larger', 'smaller'
    )
    print(f"Otro Ejemplo usando statsmodels (Test Z Unilateral, alternativa='larger') con parámetros diferentes:")
    print(f"  Tamaño del efecto (Cohen's d = δ/σ): {effect_size_statmodels_ex2:.4f}")
    print(f"  Nivel de significancia (α): {alpha_statmodels_ex2}")
    print(f"  Poder deseado (1 - β): {power_statmodels_ex2}")
    print(f"  Tamaño de muestra requerido: {math.ceil(n_statmodels_ex2)}\n")

except ImportError:
    print("La librería statsmodels no está instalada. Para usarla, ejecute: %pip install statsmodels")

print("Explicación de los Parámetros en el Cálculo del Tamaño de Muestra para Pruebas de Hipótesis:")
print("  - Nivel de Significancia (α): La probabilidad de rechazar H0 cuando es verdadera (Error Tipo I). Un α más pequeño requiere más evidencia, por lo tanto, un tamaño de muestra mayor para mantener el mismo poder.")
print("  - Poder (1 - β): La probabilidad de rechazar H0 cuando es falsa (detectar un efecto real). Un poder más alto significa una menor probabilidad de Error Tipo II (β), lo que requiere un tamaño de muestra mayor.")
print("  - Desviación Estándar (σ): La variabilidad de la población. Una mayor variabilidad hace que sea más difícil detectar una diferencia, requiriendo un tamaño de muestra mayor.")
print("  - Diferencia Mínima a Detectar (δ) o Tamaño del Efecto (d): La magnitud del efecto que el estudio está diseñado para detectar. Un efecto más pequeño requiere un tamaño de muestra mayor para ser detectado con la misma confianza y poder.")
print("  - Z_α y Z_β: Son los valores de la distribución normal estándar correspondientes a los percentiles definidos por alpha y beta, respectivamente. Se obtienen usando la función inversa de la CDF (percentile point function - ppf).")

print("\nConsideraciones adicionales para pruebas de hipótesis:")
print("- La elección de si la prueba es unilateral ('larger' o 'smaller') o bilateral ('two-sided') afecta el valor Z_α y, por lo tanto, el tamaño de muestra requerido.")
print("- Si la desviación estándar de la población (σ) es desconocida, a menudo se utiliza una estimación previa de σ (quizás de estudios piloto o literatura existente).")
```

```
# Revisar y comentar el código y las explicaciones para asegurar claridad y documentación.

# --- Sección de Introducción ---
# Introduce hypothesis testing, confidence intervals, and sample size
print("Introducción a las Pruebas de Hipótesis, Intervalos de Confianza y Tamaño de Muestra\n")
print("En el análisis de datos y la inferencia estadística, utilizamos herramientas para tomar decisiones informadas sobre una población basándonos en una muestra.")
print("Las pruebas de hipótesis son un método formal para determinar si hay suficiente evidencia en una muestra de datos para rechazar una hipótesis nula sobre una población. Nos ayudan a evaluar si una observación es estadísticamente significativa o si podría deberse simplemente al azar.")
print("Los intervalos de confianza proporcionan un rango de valores dentro del cual es probable que se encuentre un parámetro de población desconocido. Nos dan una idea de la precisión de nuestra estimación basada en la muestra.")
print("El tamaño de la muestra se refiere al número de observaciones o individuos incluidos en un estudio. Determinar un tamaño de muestra adecuado es crucial porque un tamaño de muestra insuficiente puede llevar a conclusiones poco fiables, mientras que uno excesivamente grande puede ser ineficiente en términos de recursos.")
print("\nEstos conceptos son fundamentales en la práctica porque nos permiten sacar conclusiones fiables de los datos, cuantificar la incertidumbre de nuestras estimaciones y diseñar estudios eficientes que maximicen la probabilidad de detectar efectos reales si existen.")

# --- Sección de Pruebas de Hipótesis ---
# Explicación de los Pasos 1-4 de las Pruebas de Hipótesis
print("\n---\n") # Separador para mayor claridad en el notebook
print("Pasos 1-4 de las Pruebas de Hipótesis:\n")

# Paso 1: Definir las Hipótesis
print("Paso 1: Definir las Hipótesis")
print("   - Hipótesis Nula (H0): Es la afirmación que se asume como verdadera hasta que haya evidencia suficiente en su contra. A menudo representa el status quo o la ausencia de efecto.")
print("   - Hipótesis Alternativa (H1): Es la afirmación que se desea probar. Representa lo que se concluye si se rechaza la hipótesis nula. Puede ser unilateral (mayor que o menor que) o bilateral (diferente de).\n")

# Paso 2: Establecer el Nivel de Significancia (Alpha - α)
print("Paso 2: Establecer el Nivel de Significancia (Alpha - α)")
print("   - Alpha es la probabilidad máxima de cometer un error de Tipo I (rechazar la hipótesis nula cuando es verdadera). Los niveles comunes son 0.05 (5%), 0.01 (1%) o 0.10 (10%). Un valor de alpha más bajo requiere evidencia más fuerte para rechazar H0.\n")

# Paso 3: Identificar el Estadístico de Prueba
print("Paso 3: Identificar el Estadístico de Prueba")
print("   - El estadístico de prueba es un valor calculado a partir de los datos de la muestra que se utiliza para evaluar cuánto se desvía la muestra de lo esperado bajo la hipótesis nula. La elección del estadístico de prueba depende del tipo de datos y de la pregunta de investigación (por ejemplo, estadístico z, estadístico t, estadístico chi-cuadrado, etc.).\n")

# Paso 4: Determinar el Valor Crítico y la Región de Rechazo
print("Paso 4: Determinar el Valor Crítico y la Región de Rechazo")
print("   - El valor crítico es un umbral en la distribución del estadístico de prueba que determina si se rechaza la hipótesis nula. Está basado en el nivel de significancia y la distribución del estadístico de prueba (por ejemplo, distribución normal estándar o distribución t de Student).")
print("   - La región de rechazo es el conjunto de valores del estadístico de prueba para los cuales se rechaza la hipótesis nula. Si el estadístico de prueba cae en la región de rechazo, se rechaza H0.\n")


# Ejemplo de Código para Ilustrar Conceptos de Hipótesis y Región de Rechazo:
print("Ejemplo de Código para Ilustrar Conceptos (Pasos 1-4):\n")

# Escenario hipotético: Prueba Z bilateral para la media con desviación estándar poblacional conocida.
# H0: μ = 50
# H1: μ ≠ 50 (Prueba bilateral)

# Paso 2: Nivel de significancia (alpha)
alpha = 0.05
print(f"Nivel de significancia (alpha): {alpha}")

# Paso 4: Determinar Valores Críticos y Región de Rechazo para un test bilateral Z
# Para un test Z bilateral, alpha se divide en las dos colas de la distribución.
# Usamos norm.ppf (inverse CDF) para encontrar el valor Z que deja alpha/2 en cada cola.
critical_value_lower = norm.ppf(alpha / 2)
critical_value_upper = norm.ppf(1 - alpha / 2)

print(f"Valor crítico inferior (Z) para alpha/2: {critical_value_lower:.4f}")
print(f"Valor crítico superior (Z) para 1-alpha/2: {critical_value_upper:.4f}")
print(f"Región de rechazo: Z < {critical_value_lower:.4f} o Z > {critical_value_upper:.4f}\n")


# Graficar la distribución normal estándar y sombrear la región de rechazo (Paso 4 visualización)
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1) # Distribución Normal Estándar (media=0, desv. estándar=1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Normal Estándar (Z)')

# Sombrear la región de rechazo inferior
x_reject_lower = x[x < critical_value_lower]
y_reject_lower = norm.pdf(x_reject_lower, 0, 1)
plt.fill_between(x_reject_lower, 0, y_reject_lower, color='red', alpha=0.5, label='Región de Rechazo (α/2)')

# Sombrear la región de rechazo superior
x_reject_upper = x[x > critical_value_upper]
y_reject_upper = norm.pdf(x_reject_upper, 0, 1)
plt.fill_between(x_reject_upper, 0, y_reject_upper, color='red', alpha=0.5) # No se necesita etiqueta duplicada


plt.title('Distribución Normal Estándar con Región de Rechazo (Test Bilateral)')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.axvline(critical_value_lower, color='black', linestyle='--', label=f'Valor Crítico Inferior ({critical_value_lower:.2f})')
plt.axvline(critical_value_upper, color='black', linestyle='--', label=f'Valor Crítico Superior ({critical_value_upper:.2f})')
plt.legend()
plt.grid(True)
plt.show()

print("La gráfica muestra la distribución normal estándar. Las áreas sombreadas en rojo representan la región de rechazo para un test bilateral con alpha = 0.05.")
print("Si el estadístico de prueba Z calculado a partir de los datos de la muestra cae dentro de estas áreas sombreadas,")
print(f"rechazaríamos la hipótesis nula (H0) al nivel de significancia de {alpha*100}%.")

# Explicación y Código de los Pasos 5-6 de las Pruebas de Hipótesis
print("\n---\n") # Separador
print("Pasos 5-6 de las Pruebas de Hipótesis:\n")

# Paso 5: Calcular el Estadístico de Prueba y el Valor p
print("Paso 5: Calcular el Estadístico de Prueba y el Valor p")
print("   - Calcular el estadístico de prueba utilizando los datos de la muestra. La fórmula específica depende del tipo de prueba que se esté realizando.")
print("   - El valor p es la probabilidad de obtener un estadístico de prueba tan extremo, o más extremo, como el observado en la muestra, asumiendo que la hipótesis nula es verdadera. Un valor p pequeño (típicamente menor que alpha) proporciona evidencia contra H0.\n")

# Paso 6: Tomar una Decisión y Graficar los Resultados
print("Paso 6: Tomar una Decisión y Graficar los Resultados")
print("   - Comparar el estadístico de prueba calculado con el valor crítico o comparar el valor p con el nivel de significancia (alpha).")
print("   - Si el estadístico de prueba cae en la región de rechazo (o si el valor p <= alpha), se rechaza la hipótesis nula.")
print("   - Si el estadístico de prueba no cae en la región de rechazo (o si el valor p > alpha), no se rechaza la hipótesis nula.")
print("   - Graficar la distribución, el valor crítico, el estadístico de prueba y la región de rechazo es crucial para visualizar la decisión y entender dónde cae el estadístico de prueba en relación con la región de rechazo. Esto ayuda a comunicar los resultados de la prueba de hipótesis de manera efectiva.\n")


# Simular datos de muestra y realizar cálculos (Paso 5)
# Usamos los mismos parámetros del escenario hipotético anterior.
n = 30      # Tamaño de la muestra
x_bar = 53  # Media de la muestra observada
mu_0 = 50   # Media poblacional bajo H0
sigma = 10  # Desviación estándar poblacional (supuesta conocida)

print(f"Escenario simulado: Tamaño de muestra n = {n}, Media de la muestra x̄ = {x_bar}")
print(f"Hipótesis Nula H0: μ = {mu_0}, Desviación estándar de la población σ = {sigma}\n")


# Calcular el Estadístico de Prueba Z (Paso 5)
# Fórmula: Z = (x_bar - mu_0) / (sigma / sqrt(n))
z_statistic = (x_bar - mu_0) / (sigma / np.sqrt(n))

# Calcular el valor p (para un test bilateral) (Paso 5)
# El valor p es la probabilidad de observar un estadístico Z tan extremo como el calculado, en cualquier cola.
# Usamos norm.sf (función de supervivencia = 1 - CDF) para la cola superior y multiplicamos por 2 para bilateral.
p_value = 2 * norm.sf(abs(z_statistic))


# Graficar los resultados: distribución, valores críticos, estadístico de prueba (Paso 6 visualización)
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1) # Distribución Normal Estándar

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Normal Estándar (Z)')

# Sombrear la región de rechazo (usando los valores críticos calculados previamente)
x_reject_lower = x[x < critical_value_lower]
y_reject_lower = norm.pdf(x_reject_lower, 0, 1)
plt.fill_between(x_reject_lower, 0, y_reject_lower, color='red', alpha=0.5, label='Región de Rechazo (α/2)')

x_reject_upper = x[x > critical_value_upper]
y_reject_upper = norm.pdf(x_reject_upper, 0, 1)
plt.fill_between(x_reject_upper, 0, y_reject_upper, color='red', alpha=0.5)


# Agregar líneas verticales para los valores críticos y el estadístico de prueba
plt.axvline(critical_value_lower, color='black', linestyle='--', label=f'Valor Crítico Inferior ({critical_value_lower:.2f})')
plt.axvline(critical_value_upper, color='black', linestyle='--', label=f'Valor Crítico Superior ({critical_value_upper:.2f})')
plt.axvline(z_statistic, color='green', linestyle='-', label=f'Estadístico de Prueba Z ({z_statistic:.2f})', linewidth=2)


plt.title('Distribución Normal Estándar con Región de Rechazo, Valores Críticos y Estadístico de Prueba')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

# Imprimir el estadístico de prueba calculado y el valor p
print(f"Estadístico de Prueba Z calculado: {z_statistic:.4f}")
print(f"Valor p calculado (bilateral): {p_value:.4f}\n")


# Explicación del Paso 7 de las Pruebas de Hipótesis (Decisión y Conclusión)
print("\n---\n") # Separador
print("Paso 7: Decisión Estadística y Conclusión Práctica\n")

print("Una vez que hemos calculado el estadístico de prueba y el valor p, y los hemos comparado con el valor crítico y el nivel de significancia (alpha), tomamos una decisión estadística y extraemos una conclusión práctica.\n")

print("Decisión Estadística:")
print("   - Esta decisión se basa directamente en la comparación del estadístico de prueba con el valor crítico o del valor p con alpha.")
print(f"   - Regla del Valor Crítico: Rechazamos la hipótesis nula (H0) si el estadístico de prueba cae en la región de rechazo. Para nuestro ejemplo bilateral, esto significa si Z <= {critical_value_lower:.4f} o Z >= {critical_value_upper:.4f}.")
print(f"   - Regla del Valor p: Rechazamos la hipótesis nula (H0) si el valor p es menor o igual que alpha ({alpha}).")
print("   - Si no se cumplen estas condiciones, no rechazamos la hipótesis nula (H0).\n")

print("Conclusión Práctica:")
print("   - Esta conclusión traduce la decisión estadística a un lenguaje comprensible en el contexto del problema. Debe indicar si hay suficiente evidencia para apoyar la hipótesis alternativa o si no hay suficiente evidencia para rechazar la hipótesis nula.")
print("   - Es crucial evitar la palabra 'aceptar' la hipótesis nula. No rechazar H0 no significa que H0 sea verdadera, solo que los datos de la muestra no proporcionan suficiente evidencia para rechazarla. Podría ser que la muestra no sea lo suficientemente grande o que la diferencia real sea muy pequeña para ser detectada con el tamaño de muestra actual.\n")

print("Significado e Interpretación del Valor p:")
print("   - El valor p es la probabilidad de observar los datos de la muestra (o resultados más extremos) si la hipótesis nula (H0) fuera verdadera.")
print("   - Un valor p pequeño sugiere que los datos observados son improbables bajo H0, proporcionando así evidencia contra H0.")
print(f"   - Si el valor p ({p_value:.4f}) es menor o igual a alpha ({alpha}), es improbable que los datos ocurran si H0 es cierta, por lo tanto, rechazamos H0.")
print(f"   - Si el valor p ({p_value:.4f}) es mayor que alpha ({alpha}), los datos observados son razonablemente probables si H0 es cierta, por lo tanto, no rechazamos H0.")
print("   - El valor p NO es la probabilidad de que H0 sea verdadera, ni la probabilidad de cometer un error de Tipo I.\n")

# Ejemplos de Interpretación de Resultados (basado en nuestro ejemplo anterior):
print("Ejemplos de Interpretación de Resultados (basado en nuestro ejemplo anterior):")
print(f"Nuestro estadístico de prueba Z calculado fue {z_statistic:.4f}.")
print(f"Nuestro valor p calculado fue {p_value:.4f}.")
print(f"Nuestro nivel de significancia alpha fue {alpha}.\n")

# Escenario 1: Basado en nuestros resultados actuales (no rechazar H0)
print("Escenario de nuestro Ejemplo:")
print(f"Decisión Estadística (usando el valor p): Como el valor p ({p_value:.4f}) es > alpha ({alpha}), no rechazamos la hipótesis nula.")
print(f"Decisión Estadística (usando el valor crítico): Como el estadístico de prueba Z ({z_statistic:.4f}) está entre los valores críticos ({critical_value_lower:.4f} y {critical_value_upper:.4f}), no rechazamos la hipótesis nula.")
print("Conclusión Práctica: No hay suficiente evidencia, al nivel de significancia de 5%, para concluir que la media de la población es significativamente diferente de 50. Esto no significa que la media sea exactamente 50, solo que la evidencia de esta muestra no es lo suficientemente fuerte para afirmar que es diferente.\n")

# Escenario 2: Caso hipotético donde H0 es rechazado
hypothetical_z_reject = 2.5 # Supongamos un Estadístico Z de 2.5
hypothetical_p_reject = 2 * norm.sf(abs(hypothetical_z_reject)) # Calcular el p-value para este Z hipotético
print("Escenario Hipotético (donde se rechazaría H0):")
print(f"Supongamos que el estadístico de prueba Z fuera {hypothetical_z_reject:.4f} y el valor p fuera {hypothetical_p_reject:.4f}.")
print(f"Decisión Estadística (usando el valor p): Como el valor p ({hypothetical_p_reject:.4f}) es <= alpha ({alpha}), rechazamos la hipótesis nula.")
print(f"Decisión Estadística (usando el valor crítico): Como el estadístico de prueba Z ({hypothetical_z_reject:.4f}) es >= valor crítico superior ({critical_value_upper:.4f}), rechazamos la hipótesis nula.")
print("Conclusión Práctica: Hay suficiente evidencia, al nivel de significancia de 5%, para concluir que la media de la población es significativamente diferente de 50.\n")

print("Recuerde: Una decisión de 'no rechazar H0' no prueba que H0 sea verdadera. Simplemente significa que no encontramos evidencia suficiente en la muestra para demostrar que es falsa al nivel de significancia elegido.")


# --- Sección de Intervalos de Confianza ---
print("\n---\n") # Separador
print("Intervalos de Confianza\n")

# Explicar el concepto de Intervalos de Confianza
print("Un intervalo de confianza es un rango de valores que probablemente contenga un parámetro de población desconocido (como la media o la proporción) con un cierto nivel de confianza.")
print("Proporciona una estimación de la precisión de la estimación puntual obtenida de una muestra.")
print("El propósito es cuantificar la incertidumbre asociada con una estimación de la muestra. En lugar de dar un solo número (la estimación puntual), da un rango de valores plausibles para el parámetro de la población.")
print("\nEl Nivel de Confianza:")
print("   - El nivel de confianza (comúnmente 90%, 95% o 99%) es la probabilidad de que, si se repitiera el proceso de muestreo y construcción del intervalo un gran número de veces, el intervalo resultante contenga el verdadero parámetro de la población.")
print("   - Por ejemplo, un intervalo de confianza del 95% significa que si tomáramos 100 muestras aleatorias de la misma población y calculáramos un intervalo de confianza para cada muestra, esperaríamos que aproximadamente 95 de esos intervalos contuvieran el verdadero parámetro de la población.")
print("   - No es la probabilidad de que el verdadero parámetro de la población caiga dentro de un intervalo específico calculado. El parámetro de la población es un valor fijo (aunque desconocido), mientras que el intervalo de confianza varía de una muestra a otra.")
print("   - Un nivel de confianza más alto resulta en un intervalo más amplio (mayor incertidumbre), mientras que un nivel de confianza más bajo resulta en un intervalo más estrecho (menor incertidumbre), asumiendo que todo lo demás es constante.\n")

# Calcular un Intervalo de Confianza para la Media Poblacional
# Usando los datos simulados: x_bar, sigma, n del paso anterior
# Nivel de confianza del 95%
confidence_level = 0.95
alpha_ci = 1 - confidence_level # Alpha para el intervalo de confianza
print(f"Calculando el intervalo de confianza para la media poblacional (μ) con:")
print(f"  Media muestral (x̄): {x_bar}")
print(f"  Desviación estándar poblacional (σ): {sigma}")
print(f"  Tamaño de muestra (n): {n}")
print(f"  Nivel de confianza: {confidence_level*100}%\n")


# Para desviación estándar poblacional conocida (sigma) y tamaño de muestra >= 30,
# usamos la distribución Z.
# Fórmula: x_bar ± Z * (sigma / sqrt(n))
# Z es el valor crítico de la distribución normal estándar para el nivel de confianza deseado.

# Encontrar el valor Z crítico para el nivel de confianza del 95% (bilateral)
# Usamos norm.ppf(1 - alpha_ci / 2) para encontrar el Z que deja alpha_ci/2 en la cola superior.
z_critical_ci = norm.ppf(1 - alpha_ci / 2)

# Calcular el margen de error
margin_of_error = z_critical_ci * (sigma / np.sqrt(n))

# Calcular los límites del intervalo de confianza
lower_bound_ci = x_bar - margin_of_error
upper_bound_ci = x_bar + margin_of_error

# Imprimir el Intervalo de Confianza Calculado
print(f"El valor crítico Z para un nivel de confianza del {confidence_level*100}% es aproximadamente {z_critical_ci:.4f}")
print(f"El margen de error es: {margin_of_error:.4f}")
print(f"El intervalo de confianza del {confidence_level*100}% para la media poblacional es: [{lower_bound_ci:.4f}, {upper_bound_ci:.4f}]\n")


# Usar scipy.stats.norm.interval como alternativa (verificación)
# Parámetros: nivel de confianza, loc (media), scale (error estándar = sigma / sqrt(n))
confidence_interval_scipy = norm.interval(confidence_level, loc=x_bar, scale=sigma / np.sqrt(n))

print(f"Intervalo de confianza calculado usando scipy.stats.norm.interval: {confidence_interval_scipy}\n")


# Explicar los Parámetros Utilizados en el cálculo del IC
print("Explicación de los Parámetros Utilizados para el Intervalo de Confianza:")
print(f"  - Media muestral (x̄ = {x_bar}): Es la media de los datos observados en nuestra muestra. Es la mejor estimación puntual del parámetro de la población.")
print(f"  - Desviación estándar poblacional (σ = {sigma}): Es la desviación estándar de la población completa. En escenarios reales, a menudo es desconocida y se usa la desviación estándar muestral (s) en su lugar, lo que requiere usar la distribución t de Student (especialmente para muestras pequeñas).")
print(f"  - Tamaño de muestra (n = {n}): Es el número de observaciones en nuestra muestra. Un tamaño de muestra mayor generalmente resulta en un intervalo de confianza más estrecho (más preciso) porque el error estándar (σ/√n) disminuye.")
print(f"  - Nivel de confianza ({confidence_level*100}%): Es el grado de seguridad que queremos tener de que el intervalo contenga el verdadero parámetro de la población. Un nivel más alto requiere un intervalo más amplio para proporcionar esa mayor seguridad.")
print(f"  - Valor crítico Z ({z_critical_ci:.4f}): Es el número de errores estándar que nos alejamos de la media muestral para construir el intervalo. Depende del nivel de confianza y de la distribución utilizada (Z en este caso).")

print("\nCómo afectan estos parámetros al intervalo:")
print("  - x̄ determina el centro del intervalo.")
print("  - σ y n afectan el ancho del intervalo a través del error estándar (σ/√n). Un σ más grande o un n más pequeño aumentan el error estándar y hacen el intervalo más amplio.")
print("  - El nivel de confianza afecta el valor crítico Z. Un nivel de confianza más alto requiere un Z más grande, lo que aumenta el margen de error y hace el intervalo más amplio.\n")


# Interpretar el Intervalo de Confianza Calculado
print("Interpretación del Intervalo de Confianza:")
print(f"Basado en nuestra muestra, tenemos un {confidence_level*100}% de confianza en que el verdadero valor de la media poblacional (μ) se encuentra entre {lower_bound_ci:.4f} y {upper_bound_ci:.4f}.")
print("Esto significa que si repitiéramos este estudio muchas veces, el 95% de los intervalos de confianza construidos de esta manera contendrían la verdadera media poblacional.")
print("No podemos decir que hay un 95% de probabilidad de que la verdadera media poblacional caiga dentro de *este* intervalo específico, ya que la verdadera media es un valor fijo y o bien está en el intervalo o no lo está. La probabilidad se refiere al método de construcción del intervalo.")
print("El intervalo nos da un rango de valores plausibles para la media poblacional, dada la evidencia de nuestra muestra.")


# --- Sección de Tamaño de Muestra ---
print("\n---\n") # Separador
print("Concepto de Tamaño de Muestra y su Cálculo\n")

# Explicar el concepto de tamaño de muestra y su importancia
print("El tamaño de la muestra (n) es simplemente el número de individuos u observaciones incluidas en un estudio estadístico.")
print("Su importancia es fundamental por varias razones:")
print("   - Precisión de las Estimaciones: Un tamaño de muestra más grande generalmente conduce a estimaciones más precisas de los parámetros de la población (como la media o la proporción). Con más datos, la estimación puntual de la muestra se acerca más al verdadero valor del parámetro poblacional.")
print("   - Poder de las Pruebas de Hipótesis: Un tamaño de muestra adecuado aumenta la 'potencia' de una prueba de hipótesis, que es la probabilidad de detectar un efecto o diferencia real si existe (es decir, rechazar correctamente una hipótesis nula falsa). Un tamaño de muestra insuficiente puede llevar a no detectar un efecto real (error de Tipo II).")
print("   - Generalización de los Resultados: Un tamaño de muestra representativo y lo suficientemente grande permite generalizar los hallazgos de la muestra a la población de la que se extrajo.")
print("   - Eficiencia de los Recursos: Si bien un tamaño de muestra más grande mejora la precisión y la potencia, también requiere más recursos (tiempo, dinero, esfuerzo). Determinar el tamaño de muestra necesario ayuda a optimizar el uso de estos recursos sin comprometer la validez del estudio.")
print("\nEn resumen, elegir un tamaño de muestra apropiado es un paso crítico en el diseño de un estudio para garantizar que los resultados sean fiables, generalizables y que la probabilidad de sacar conclusiones erróneas se minimice.")

# Explicar los factores que influyen en el tamaño de muestra
print("\nFactores que Influyen en el Tamaño de Muestra Requerido\n")
print("Varios factores clave determinan el tamaño de muestra necesario para un estudio estadístico:")
print("   - Margen de Error Deseado (E): Es la cantidad máxima de diferencia que estás dispuesto a tolerar entre la estimación de la muestra y el verdadero valor del parámetro de la población. Un margen de error más pequeño requiere un tamaño de muestra más grande para lograr una mayor precisión.")
print("   - Nivel de Confianza Deseado (1 - α): Es la probabilidad de que el intervalo de confianza contenga el verdadero parámetro de la población si el proceso se repite muchas veces. Un nivel de confianza más alto (por ejemplo, 99% en lugar de 95%) requiere un tamaño de muestra más grande.")
print("   - Variabilidad de la Población (σ o s): Se refiere a la dispersión de los datos en la población, típicamente medida por la desviación estándar (σ para la población, s para la muestra). Una mayor variabilidad en la población requiere un tamaño de muestra más grande para obtener estimaciones precisas.")
print("   - Poder Deseado (1 - β): En el contexto de las pruebas de hipótesis, el poder es la probabilidad de detectar un efecto o diferencia real cuando existe. Un poder más alto (por ejemplo, 90% en lugar de 80%) requiere un tamaño de muestra más grande.")
print("   - Nivel de Significancia (α): En el contexto de las pruebas de hipótesis, alpha es la probabilidad de cometer un error de Tipo I (rechazar H0 cuando es verdadera). Un alpha más pequeño (por ejemplo, 0.01 en lugar de 0.05) puede indirectamente requerir un tamaño de muestra más grande para mantener el poder.")
print("   - Tamaño del Efecto (d): Para pruebas de hipótesis, es la magnitud de la diferencia o relación que se espera detectar. Un tamaño del efecto más pequeño (es decir, una diferencia más sutil) requiere un tamaño de muestra más grande para ser detectado con suficiente poder.")


# Calcular tamaño de muestra para estimar una media poblacional
print("\nCálculo del Tamaño de Muestra para Estimar una Media Poblacional\n")

# Fórmula utilizada para estimar la media: n = (Z * σ / E)^2
print("Fórmula utilizada: n = (Z * σ / E)^2")
print("Donde:")
print("  n = Tamaño de muestra requerido")
print("  Z = Puntuación Z para el nivel de confianza deseado (por ejemplo, 1.96 para 95% de confianza)")
print("  σ = Desviación estándar de la población (o una estimación previa)")
print("  E = Margen de error máximo aceptable\n")


# Ejemplo de Cálculo 1: Estimar media con 95% confianza, margen de error 5, desv. estándar 10
desired_confidence_level = 0.95
desired_margin_of_error = 5
estimated_sigma = 10 # Necesitamos una estimación de la desviación estándar poblacional

# Encontrar la puntuación Z para el nivel de confianza deseado (bilateral)
alpha_ss = 1 - desired_confidence_level
z_score_estimate = norm.ppf(1 - alpha_ss / 2) # Z para el percentil (1 - alpha/2)

# Calcular el tamaño de muestra requerido
n_estimate_mean = math.ceil((z_score_estimate * estimated_sigma / desired_margin_of_error) ** 2) # math.ceil redondea hacia arriba al entero más cercano

print(f"Ejemplo 1: Estimar la media poblacional")
print(f"  Nivel de confianza deseado: {desired_confidence_level*100}%")
print(f"  Margen de error deseado (E): {desired_margin_of_error}")
print(f"  Desviación estándar estimada (σ): {estimated_sigma}")
print(f"  Puntuación Z para {desired_confidence_level*100}%: {z_score_estimate:.4f}")
print(f"  Tamaño de muestra requerido: ({z_score_estimate:.4f} * {estimated_sigma} / {desired_margin_of_error})^2 = {n_estimate_mean}\n")


# Ejemplo de Cálculo 2: Parámetros diferentes
desired_confidence_level_ex2 = 0.99
desired_margin_of_error_ex2 = 2
estimated_sigma_ex2 = 15

alpha_ss_ex2 = 1 - desired_confidence_level_ex2
z_score_estimate_ex2 = norm.ppf(1 - alpha_ss_ex2 / 2)
n_estimate_mean_ex2 = math.ceil((z_score_estimate_ex2 * estimated_sigma_ex2 / desired_margin_of_error_ex2) ** 2)

print(f"Ejemplo 2: Estimar la media poblacional con parámetros diferentes")
print(f"  Nivel de confianza deseado: {desired_confidence_level_ex2*100}%")
print(f"  Margen de error deseado (E): {desired_margin_of_error_ex2}")
print(f"  Desviación estándar estimada (σ): {estimated_sigma_ex2}")
print(f"  Puntuación Z para {desired_confidence_level_ex2*100}%: {z_score_estimate_ex2:.4f}")
print(f"  Tamaño de muestra requerido: ({z_score_estimate_ex2:.4f} * {estimated_sigma_ex2} / {desired_margin_of_error_ex2})^2 = {n_estimate_mean_ex2}\n")

print("Observaciones:")
print("- Un margen de error más pequeño (Ejemplo 2 vs Ejemplo 1) requiere un tamaño de muestra significativamente mayor.")
print("- Un nivel de confianza más alto (Ejemplo 2 vs Ejemplo 1) requiere un tamaño de muestra mayor.")
print("- Una mayor variabilidad (σ) en la población también requiere un tamaño de muestra mayor (comparando si σ fuera diferente en los ejemplos con otros parámetros iguales).")


# Calcular tamaño de muestra para una prueba de hipótesis (Test Z Unilateral para la Media)
print("Cálculo del Tamaño de Muestra para una Prueba de Hipótesis (Test Z Unilateral para la Media)\n")

# Cálculo para un test Z unilateral para la media. Requiere:
# - Nivel de significancia (alpha)
# - Poder deseado (1 - beta)
# - Tamaño del efecto (delta / sigma), donde delta es la diferencia mínima a detectar.

# Fórmula para tamaño de muestra (test unilateral, sigma conocida):
# n = ((Z_alpha + Z_beta) * σ / δ)^2
print("Fórmula utilizada (Test Unilateral): n = ((Z_α + Z_β) * σ / δ)^2")
print("Donde:")
print("  n = Tamaño de muestra requerido")
print("  Z_α = Puntuación Z para el nivel de significancia (α)")
print("  Z_β = Puntuación Z para el poder deseado (1 - β). Beta (β) es la probabilidad de un error Tipo II.")
print("  σ = Desviación estándar de la población (o una estimación previa)")
print("  δ = Diferencia mínima que se desea detectar (Tamaño del efecto en unidades originales)\n")

# Nota: Para un test bilateral, la fórmula usa Z_(alpha/2).
# La fórmula también se puede expresar en términos de tamaño del efecto estandarizado (Cohen's d = δ / σ):
# n = ((Z_alpha + Z_beta) / d)^2
print("Alternativamente, usando el tamaño del efecto estandarizado (d = δ / σ): n = ((Z_α + Z_β) / d)^2\n")


# Ejemplo de Cálculo: Test Z Unilateral
# Queremos detectar una diferencia de 3 unidades de la media nula (e.g., H0: mu=50, queremos detectar si mu=53 o más)
# Nivel de significancia (alpha): 0.05 (test unilateral)
# Poder deseado: 0.80 (beta = 0.20)
# Desviación estándar poblacional estimada (sigma): 10
# Diferencia mínima a detectar (delta): 3

significance_level = 0.05
desired_power = 0.80
estimated_sigma = 10
minimum_detectable_difference = 3 # delta

# Encontrar la puntuación Z para alpha (para un test unilateral, usamos 1-alpha)
z_alpha = norm.ppf(1 - significance_level)

# Encontrar la puntuación Z para beta (para calcular el poder, usamos 1-beta)
beta = 1 - desired_power
z_beta = norm.ppf(1 - beta) # Este es el Z-score correspondiente al percentil del poder deseado

# Calcular el tamaño de muestra requerido
n_hypothesis_test = math.ceil(((z_alpha + z_beta) * estimated_sigma / minimum_detectable_difference) ** 2)

print(f"Ejemplo: Prueba de Hipótesis (Test Z Unilateral)")
print(f"  Nivel de significancia (α): {significance_level}")
print(f"  Poder deseado (1 - β): {desired_power}")
print(f"  Desviación estándar estimada (σ): {estimated_sigma}")
print(f"  Diferencia mínima a detectar (δ): {minimum_detectable_difference}")
print(f"  Puntuación Z para α ({significance_level}): {z_alpha:.4f}")
print(f"  Puntuación Z para 1-β ({desired_power}): {z_beta:.4f}")
print(f"  Tamaño de muestra requerido: (({z_alpha:.4f} + {z_beta:.4f}) * {estimated_sigma} / {minimum_detectable_difference})^2 = {n_hypothesis_test}\n")


# Otro Ejemplo: Parámetros diferentes
# Nivel de significancia (alpha): 0.01 (unilateral)
# Poder deseado: 0.95
# Desviación estándar poblacional estimada (sigma): 15
# Diferencia mínima a detectar (delta): 2

significance_level_ex2 = 0.01
desired_power_ex2 = 0.95
estimated_sigma_ex2 = 15
minimum_detectable_difference_ex2 = 2

z_alpha_ex2 = norm.ppf(1 - significance_level_ex2)
beta_ex2 = 1 - desired_power_ex2
z_beta_ex2 = norm.ppf(1 - beta_ex2)
n_hypothesis_test_ex2 = math.ceil(((z_alpha_ex2 + z_beta_ex2) * estimated_sigma_ex2 / minimum_detectable_difference_ex2) ** 2)


print(f"Otro Ejemplo: Prueba de Hipótesis (Test Z Unilateral) con parámetros diferentes")
print(f"  Nivel de significancia (α): {significance_level_ex2}")
print(f"  Poder deseado (1 - β): {desired_power_ex2}")
print(f"  Desviación estándar estimada (σ): {estimated_sigma_ex2}")
print(f"  Diferencia mínima a detectar (δ): {minimum_detectable_difference_ex2}")
print(f"  Puntuación Z para α ({significance_level_ex2}): {z_alpha_ex2:.4f}")
print(f"  Puntuación Z para 1-β ({desired_power_ex2}): {z_beta_ex2:.4f}")
print(f"  Tamaño de muestra requerido: (({z_alpha_ex2:.4f} + {z_beta_ex2:.4f}) * {estimated_sigma_ex2} / {minimum_detectable_difference_ex2})^2 = {n_hypothesis_test_ex2}\n")

print("Observaciones:")
print("- Un poder deseado más alto (Ejemplo 2 vs Ejemplo 1) requiere un tamaño de muestra mayor.")
print("- Un nivel de significancia más estricto (alpha más pequeño, Ejemplo 2 vs Ejemplo 1) requiere un tamaño de muestra mayor.")
print("- La detección de una diferencia más pequeña (delta más pequeño, Ejemplo 2 vs Ejemplo 1) requiere un tamaño de muestra significativamente mayor.")
print("- Una mayor variabilidad (σ) también requiere un tamaño de muestra mayor.")


# Usando statsmodels para análisis de poder (alternativa más completa)
# Nota: statsmodels podría necesitar instalación.
# %pip install statsmodels # Descomentar y ejecutar si es necesario

try:
    # Asegurarse de que statsmodels está importado
    from statsmodels.stats.power import NormalIndPower
    # from statsmodels.stats.proportion import proportion_effectsize # No se usa en estos ejemplos de media

    print("Cálculo del Tamaño de Muestra usando statsmodels (Alternativa)\n")

    # Ejemplo usando statsmodels para test Z unilateral para la media
    # Necesitamos especificar el tamaño del efecto estandarizado (Cohen's d = delta / sigma)
    # Para el primer ejemplo: delta=3, sigma=10, entonces Cohen's d = 3/10 = 0.3 (tamaño del efecto medio)

    effect_size_statmodels = minimum_detectable_difference / estimated_sigma
    alpha_statmodels = significance_level
    power_statmodels = desired_power
    ratio = 0 # ratio de tamaños de muestra, 0 para test de una muestra

    # Crear un objeto de análisis de poder para distribución normal (test Z)
    power_analysis = NormalIndPower()

    # Calcular tamaño de muestra (por grupo, pero para test de una muestra es el total)
    # 'solve_for' = 'nobs1' calcula el tamaño de muestra para el primer grupo (nuestra única muestra)
    # 'alternative' especifica el tipo de test ('two-sided', 'larger' o 'smaller')
    n_statmodels = power_analysis.solve_power(
        effect_size=effect_size_statmodels,
        alpha=alpha_statmodels,
        power=power_statmodels,
        ratio=ratio, # Usar ratio=0 para test de una muestra
        alternative='larger' # Opciones válidas: 'two-sided', 'larger', 'smaller'. Usamos 'larger' asumiendo el ejemplo unilateral.
    )

    print(f"Ejemplo usando statsmodels (Test Z Unilateral para la Media, alternativa='larger'):")
    print(f"  Tamaño del efecto (Cohen's d = δ/σ): {effect_size_statmodels:.4f}")
    print(f"  Nivel de significancia (α): {alpha_statmodels}")
    print(f"  Poder deseado (1 - β): {power_statmodels}")
    print(f"  Tamaño de muestra requerido: {math.ceil(n_statmodels)}\n")


    # Ejemplo usando statsmodels para el segundo escenario
    effect_size_statmodels_ex2 = minimum_detectable_difference_ex2 / estimated_sigma_ex2
    alpha_statmodels_ex2 = significance_level_ex2
    power_statmodels_ex2 = desired_power_ex2

    n_statmodels_ex2 = power_analysis.solve_power(
        effect_size=effect_size_statmodels_ex2,
        alpha=alpha_statmodels_ex2,
        power=power_statmodels_ex2,
        ratio=ratio,
        alternative='larger' # Usamos 'larger' asumiendo el ejemplo unilateral.
    )
    print(f"Otro Ejemplo usando statsmodels (Test Z Unilateral, alternativa='larger') con parámetros diferentes:")
    print(f"  Tamaño del efecto (Cohen's d = δ/σ): {effect_size_statmodels_ex2:.4f}")
    print(f"  Nivel de significancia (α): {alpha_statmodels_ex2}")
    print(f"  Poder deseado (1 - β): {power_statmodels_ex2}")
    print(f"  Tamaño de muestra requerido: {math.ceil(n_statmodels_ex2)}\n")

except ImportError:
    print("La librería statsmodels no está instalada. Para usarla, ejecute: %pip install statsmodels")

print("Explicación de los Parámetros en el Cálculo del Tamaño de Muestra para Pruebas de Hipótesis:")
print("  - Nivel de Significancia (α): La probabilidad de rechazar H0 cuando es verdadera (Error Tipo I). Un α más pequeño requiere más evidencia, por lo tanto, un tamaño de muestra mayor para mantener el mismo poder.")
print("  - Poder (1 - β): La probabilidad de rechazar H0 cuando es falsa (detectar un efecto real). Un poder más alto significa una menor probabilidad de Error Tipo II (β), lo que requiere un tamaño de muestra mayor.")
print("  - Desviación Estándar (σ): La variabilidad de la población. Una mayor variabilidad hace que sea más difícil detectar una diferencia, requiriendo un tamaño de muestra mayor.")
print("  - Diferencia Mínima a Detectar (δ) o Tamaño del Efecto (d): La magnitud del efecto que el estudio está diseñado para detectar. Un efecto más pequeño requiere un tamaño de muestra mayor para ser detectado con la misma confianza y poder.")
print("  - Z_α y Z_β: Son los valores de la distribución normal estándar correspondientes a los percentiles definidos por alpha y beta, respectivamente. Se obtienen usando la función inversa de la CDF (percentile point function - ppf).")

print("\nConsideraciones adicionales para pruebas de hipótesis:")
print("- La elección de si la prueba es unilateral ('larger' o 'smaller') o bilateral ('two-sided') afecta el valor Z_α y, por lo tanto, el tamaño de muestra requerido.")
print("- Si la desviación estándar de la población (σ) es desconocida, a menudo se utiliza una estimación previa de σ (quizás de estudios piloto o literatura existente).")
print("- statsmodels es una herramienta poderosa que puede calcular el tamaño de muestra para una variedad de pruebas estadísticas y escenarios (por ejemplo, pruebas t, pruebas de proporciones, ANOVA), ofreciendo más flexibilidad que las fórmulas manuales directas.")
```

```
from scipy.stats import norm
import math

# --- Cálculo Específico de Tamaño de Muestra (Estimación de Proporción) ---
print("\n--- Cálculo Específico de Tamaño de Muestra (Estimación de Proporción) ---\n")

# Fórmula para tamaño de muestra para estimar una proporción:
# n = (Z^2 * p * (1-p)) / E^2
# Donde:
# n = tamaño de muestra requerido
# Z = Puntuación Z para el nivel de confianza deseado
# p = proporción estimada de la población (si se conoce, de lo contrario usar 0.5)
# E = margen de error máximo aceptable

print("Fórmula utilizada: n = (Z^2 * p * (1-p)) / E^2")
print("Donde:")
print("  n = Tamaño de muestra requerido")
print("  Z = Puntuación Z para el nivel de confianza deseado")
print("  p = Proporción estimada de la población (usar 0.5 si no se tiene una estimación previa)")
print("  E = Margen de error máximo aceptable\n")

# Datos proporcionados por el usuario (Ejemplo de Residentes Vacunados):
desired_margin_of_error_prop = 0.03 # Margen de error del 3%
confidence_level_prop = 0.95      # Nivel de confianza del 95%

print(f"Datos del problema (Ejemplo de Residentes Vacunados):")
print(f"  Margen de error deseado (E): {desired_margin_of_error_prop*100}%")
print(f"  Nivel de confianza: {confidence_level_prop*100}%\n")

# Encontrar la puntuación Z para el nivel de confianza deseado (bilateral)
alpha_prop = 1 - confidence_level_prop
z_score_prop = norm.ppf(1 - alpha_prop / 2) # Z para el percentil (1 - alpha/2)

print(f"Puntuación Z para {confidence_level_prop*100}% de confianza: {z_score_prop:.4f}\n")


# Escenario 1: Con estudio piloto (proporción estimada p = 0.72)
estimated_proportion_pilot = 0.72
print(f"Escenario 1: Con estudio piloto (proporción estimada p = {estimated_proportion_pilot})")

# Calcular el tamaño de muestra requerido con la estimación previa
n_prop_pilot = math.ceil((z_score_prop**2 * estimated_proportion_pilot * (1 - estimated_proportion_pilot)) / (desired_margin_of_error_prop**2))

print(f"  Tamaño de muestra requerido: ({z_score_prop:.4f}^2 * {estimated_proportion_pilot} * {1 - estimated_proportion_pilot}) / {desired_margin_of_error_prop}^2 = {n_prop_pilot}\n")


# Escenario 2: Sin estudio piloto (proporción estimada p = 0.5)
# Cuando no se tiene una estimación previa, se usa p = 0.5 para maximizar el tamaño de muestra
# y garantizar que el margen de error no exceda el deseado, independientemente de la verdadera proporción.
estimated_proportion_no_pilot = 0.5
print(f"Escenario 2: Sin estudio piloto (usando p = {estimated_proportion_no_pilot} para máxima variabilidad)")

# Calcular el tamaño de muestra requerido sin estimación previa
n_prop_no_pilot = math.ceil((z_score_prop**2 * estimated_proportion_no_pilot * (1 - estimated_proportion_no_pilot)) / (desired_margin_of_error_prop**2))

print(f"  Tamaño de muestra requerido: ({z_score_prop:.4f}^2 * {estimated_proportion_no_pilot} * {1 - estimated_proportion_no_pilot}) / {desired_margin_of_error_prop}^2 = {n_prop_no_pilot}\n")


print("Observaciones (Ejemplo de Residentes Vacunados):")
print(f"- El tamaño de muestra es mayor ({n_prop_no_pilot}) cuando no se tiene una estimación previa (p=0.5)")
print(f"  en comparación con tener una estimación previa ({n_prop_pilot}), ya que p=0.5 maximiza el valor de p*(1-p).")
print("- Esto asegura que el tamaño de muestra sea suficiente incluso if the verdadera proporción está cerca de 0.5.")


# --- Nuevo Cálculo Específico de Tamaño de Muestra (Estimación de Proporción - Apoyo a Política) ---
print("\n--- Cálculo Específico de Tamaño de Muestra (Estimación de Proporción - Apoyo a Política) ---\n")

# Datos proporcionados por el usuario:
informal_proportion_support = 0.56 # Proporción informal de apoyo (p)
desired_margin_of_error_support = 0.05 # Margen de error deseado (E)
confidence_level_support = 0.95      # Nivel de confianza deseado

print(f"Datos del problema:")
print(f"  Proporción informal estimada (p): {informal_proportion_support}")
print(f"  Margen de error deseado (E): {desired_margin_of_error_support*100}%")
print(f"  Nivel de confianza: {confidence_level_support*100}%\n")

# Encontrar la puntuación Z para el nivel de confianza deseado (bilateral)
# Para un nivel de confianza del 95%, alpha = 0.05. Z es el valor que deja alpha/2 = 0.025 en cada cola.
alpha_support = 1 - confidence_level_support
z_score_support = norm.ppf(1 - alpha_support / 2) # Z para el percentil (1 - alpha/2)

print(f"Puntuación Z para {confidence_level_support*100}% de confianza: {z_score_support:.4f}\n")

# Calcular el tamaño de muestra requerido con la estimación previa informal
n_support_informal = math.ceil((z_score_support**2 * informal_proportion_support * (1 - informal_proportion_support)) / (desired_margin_of_error_support**2))

print(f"Cálculo del tamaño de muestra requerido (con estimación informal):")
print(f"  n = ({z_score_support:.4f}^2 * {informal_proportion_support} * {1 - informal_proportion_support}) / {desired_margin_of_error_support}^2")
print(f"  Tamaño de muestra requerido: {n_support_informal}\n")

# Escenario sin datos preliminares (p = 0.5)
estimated_proportion_no_prelim = 0.5
print(f"Cálculo del tamaño de muestra requerido (sin datos preliminares, usando p = {estimated_proportion_no_prelim}):")

# Calcular el tamaño de muestra requerido sin datos preliminares
n_support_no_prelim = math.ceil((z_score_support**2 * estimated_proportion_no_prelim * (1 - estimated_proportion_no_prelim)) / (desired_margin_of_error_support**2))

print(f"  n = ({z_score_support:.4f}^2 * {estimated_proportion_no_prelim} * {1 - estimated_proportion_no_prelim}) / {desired_margin_of_error_support}^2")
print(f"  Tamaño de muestra requerido: {n_support_no_prelim}\n")


print("Justificación para usar p = 0.5 cuando no hay datos preliminares:")
print("- La fórmula para el tamaño de muestra para una proporción es n = (Z^2 * p * (1-p)) / E^2.")
print("- El término p*(1-p) representa la variabilidad de la proporción.")
print("- El producto p*(1-p) alcanza su valor máximo cuando p = 0.5. En este caso, p*(1-p) = 0.5 * 0.5 = 0.25.")
print("- Al usar p = 0.5 en la fórmula, se calcula el tamaño de muestra más grande posible para un Z y E dados.")
print("- Esto garantiza que el tamaño de muestra calculado sea suficiente para lograr el margen de error deseado con el nivel de confianza especificado, sin importar cuál sea la verdadera proporción poblacional. Es un enfoque conservador que asegura una muestra lo suficientemente grande.")
print(f"\nEn este ejemplo, sin datos preliminares, el tamaño de muestra requerido sería {n_support_no_prelim}, que es mayor que el tamaño de muestra requerido ({n_support_informal}) cuando se utiliza la estimación informal de p=0.56.")


# --- Nuevo Cálculo Específico de Tamaño de Muestra (Estimación de Proporción - Pacientes) ---
print("\n--- Cálculo Específico de Tamaño de Muestra (Estimación de Proporción - Pacientes) ---\n")

# Datos proporcionados por el usuario:
preliminary_proportion_patients = 0.69 # Proporción preliminar de mejoría (p)
desired_margin_of_error_patients = 0.04 # Margen de error deseado (E)
confidence_level_patients = 0.95      # Nivel de confianza deseado

print(f"Datos del problema:")
print(f"  Proporción preliminar estimada (p): {preliminary_proportion_patients}")
print(f"  Margen de error deseado (E): {desired_margin_of_error_patients*100}%")
print(f"  Nivel de confianza: {confidence_level_patients*100}%\n")

# Encontrar la puntuación Z para el nivel de confianza deseado (bilateral)
alpha_patients = 1 - confidence_level_patients
z_score_patients = norm.ppf(1 - alpha_patients / 2) # Z para el percentil (1 - alpha/2)

print(f"Puntuación Z para {confidence_level_patients*100}% de confianza: {z_score_patients:.4f}\n")

# Escenario 1: Con prueba preliminar (proporción estimada p = 0.69)
print(f"Escenario 1: Con prueba preliminar (proporción estimada p = {preliminary_proportion_patients})")

# Calcular el tamaño de muestra requerido con la estimación previa
n_patients_prelim = math.ceil((z_score_patients**2 * preliminary_proportion_patients * (1 - preliminary_proportion_patients)) / (desired_margin_of_error_patients**2))

print(f"  Tamaño de muestra requerido: ({z_score_patients:.4f}^2 * {preliminary_proportion_patients} * {1 - preliminary_proportion_patients}) / {desired_margin_of_error_patients}^2 = {n_patients_prelim}\n")


# Escenario 2: Conservador (usando p = 0.5)
estimated_proportion_conservative = 0.5
print(f"Escenario 2: Conservador (usando p = {estimated_proportion_conservative} para máxima variabilidad)")

# Calcular el tamaño de muestra requerido con el enfoque conservador
n_patients_conservative = math.ceil((z_score_patients**2 * estimated_proportion_conservative * (1 - estimated_proportion_conservative)) / (desired_margin_of_error_patients**2))

print(f"  Tamaño de muestra requerido: ({z_score_patients:.4f}^2 * {estimated_proportion_conservative} * {1 - estimated_proportion_conservative}) / {desired_margin_of_error_patients}^2 = {n_patients_conservative}\n")


print("Observaciones (Ejemplo de Pacientes):")
print(f"- El tamaño de muestra conservador ({n_patients_conservative}) es mayor que el requerido con la estimación preliminar ({n_patients_prelim}).")
print("- Usar p=0.5 garantiza que el tamaño de muestra sea suficiente, incluso si la proporción real es diferente de la estimación preliminar.")
```

```
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

# --- Prueba de Hipótesis Específica (Test Z Unilateral Inferior para la Media - Cereal) ---
print("\n--- Prueba de Hipótesis Específica (Test Z Unilateral Inferior para la Media - Cereal) ---\n")

# Datos del problema:
mu_0_cereal = 500      # Media poblacional bajo H0 (g)
sample_mean_cereal = 495   # Media de la muestra (g)
sample_size_cereal = 40    # Tamaño de la muestra
pop_std_dev_cereal = 15    # Desviación estándar poblacional (σ) en g
alpha_cereal = 0.05      # Nivel de significancia

print(f"Datos del problema:")
print(f"  Media poblacional bajo H0 (μ₀): {mu_0_cereal} g")
print(f"  Media muestral (x̄): {sample_mean_cereal} g")
print(f"  Tamaño de muestra (n): {sample_size_cereal}")
print(f"  Desviación estándar poblacional (σ): {pop_std_dev_cereal} g")
print(f"  Nivel de significancia (α): {alpha_cereal}\n")


# 1. Plantear las hipótesis nula y alternativa
print("1. Planteamiento de Hipótesis:")
print(f"   Hipótesis Nula (H₀): μ = {mu_0_cereal} g (La máquina está calibrada)")
# La sospecha es que la máquina está subllenando, por lo tanto, la media es menor que 500g.
print(f"   Hipótesis Alternativa (H₁): μ < {mu_0_cereal} g (La máquina está subllenando - Test unilateral inferior)\n")


# 2. Calcular el estadístico Z
# Fórmula para Test Z de una muestra con sigma conocida: Z = (x_bar - mu_0) / (sigma / sqrt(n))
standard_error_cereal = pop_std_dev_cereal / np.sqrt(sample_size_cereal)
z_statistic_cereal = (sample_mean_cereal - mu_0_cereal) / standard_error_cereal

print("2. Cálculo del Estadístico de Prueba Z:")
print(f"   Error estándar: σ / √n = {pop_std_dev_cereal} / √{sample_size_cereal} = {standard_error_cereal:.4f} g")
print(f"   Estadístico Z = (x̄ - μ₀) / Error estándar = ({sample_mean_cereal} - {mu_0_cereal}) / {standard_error_cereal:.4f} = {z_statistic_cereal:.4f}\n")


# 3. Determinar la región crítica y la región de rechazo.
# Para un test unilateral inferior con nivel de significancia alpha, el valor crítico Z es el valor
# que deja 'alpha' área en la cola inferior de la distribución normal estándar.
critical_value_cereal = norm.ppf(alpha_cereal)

print("3. Determinación del Valor Crítico y la Región de Rechazo:")
print(f"   Nivel de significancia (α): {alpha_cereal}")
print(f"   Para un test unilateral inferior, el valor crítico Z es norm.ppf(α).")
print(f"   Valor crítico Z: {critical_value_cereal:.4f}")
print(f"   Región de rechazo: Z < {critical_value_cereal:.4f}\n")


# Graficar la distribución normal estándar, región de rechazo, valor crítico y estadístico Z
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1) # Distribución Normal Estándar

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Normal Estándar (Z)')

# Sombrear la región de rechazo (cola inferior)
x_reject_cereal = x[x < critical_value_cereal]
y_reject_cereal = norm.pdf(x_reject_cereal, 0, 1)
plt.fill_between(x_reject_cereal, 0, y_reject_cereal, color='red', alpha=0.5, label='Región de Rechazo (α)')

# Agregar líneas verticales para el valor crítico y el estadístico de prueba
plt.axvline(critical_value_cereal, color='black', linestyle='--', label=f'Valor Crítico Z ({critical_value_cereal:.2f})')
plt.axvline(z_statistic_cereal, color='green', linestyle='-', label=f'Estadístico de Prueba Z ({z_statistic_cereal:.2f})', linewidth=2)


plt.title('Distribución Normal Estándar con Región de Rechazo y Estadístico de Prueba (Test Unilateral Inferior)')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# Calcular el valor p (para un test unilateral inferior)
# El valor p es la probabilidad de obtener un estadístico Z tan pequeño o más pequeño que el observado.
p_value_cereal = norm.cdf(z_statistic_cereal)

print(f"Valor p calculado (unilateral inferior): {p_value_cereal:.4f}\n")


# 4. Concluir si se rechaza H0.
print("4. Decisión Estadística y Conclusión:")

# Comparar el estadístico de prueba con el valor crítico O el valor p con alpha
print(f"Comparando el Estadístico Z ({z_statistic_cereal:.4f}) con el Valor Crítico Z ({critical_value_cereal:.4f}):")
if z_statistic_cereal < critical_value_cereal:
    print("   El Estadístico Z cae en la región de rechazo.")
    decision = "Rechazar H₀"
else:
    print("   El Estadístico Z NO cae en la región de rechazo.")
    decision = "No rechazar H₀"

print(f"\nComparando el Valor p ({p_value_cereal:.4f}) con α ({alpha_cereal}):")
if p_value_cereal <= alpha_cereal:
     print(f"   El Valor p ({p_value_cereal:.4f}) es menor o igual que α ({alpha_cereal}).")
     decision_p_value = "Rechazar H₀"
else:
     print(f"   El Valor p ({p_value_cereal:.4f}) es mayor que α ({alpha_cereal}).")
     decision_p_value = "No rechazar H₀"

print(f"\nDecisión Estadística: {decision}")

print("\nConclusión Práctica:")
if decision == "Rechazar H₀":
    print(f"   Al nivel de significancia del {alpha_cereal*100}%, hay suficiente evidencia para rechazar la hipótesis nula.")
    print(f"   Concluimos que el peso promedio de los paquetes de cereal es significativamente menor a {mu_0_cereal} g,")
    print("   sugiriendo que la máquina podría estar descalibrada y subllenando.")
else:
    print(f"   Al nivel de significancia del {alpha_cereal*100}%, no hay suficiente evidencia para rechazar la hipótesis nula.")
    print(f"   No podemos concluir que el peso promedio de los paquetes de cereal sea significativamente menor a {mu_0_cereal} g.")
    print("   La diferencia observada de 495 g podría deberse a la variación aleatoria de la muestra.")

print("\nInterpretación del Valor p:")
print(f"   El valor p ({p_value_cereal:.4f}) es la probabilidad de observar una media muestral de 495 g o menos,")
print(f"   si la verdadera media poblacional fuera 500 g (es decir, si H₀ fuera verdadera).")
print(f"   Como {p_value_cereal:.4f} > {alpha_cereal}, esta probabilidad es relativamente alta bajo H₀,")
print("   por lo que los datos no proporcionan evidencia fuerte contra H₀.")
```

```
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

# --- Prueba de Hipótesis Específica (Test Z Bilateral para la Media - Examen Nacional) ---
print("\n--- Prueba de Hipótesis Específica (Test Z Bilateral para la Media - Examen Nacional) ---\n")

# Datos del problema:
mu_0_exam_ht = 75      # Media poblacional bajo H0 (promedio nacional)
sample_mean_exam_ht = 78   # Media de la muestra (promedio de la clase)
sample_size_exam_ht = 36    # Tamaño de la muestra
pop_std_dev_exam_ht = 12    # Desviación estándar poblacional (σ)
alpha_exam_ht = 0.05      # Nivel de significancia

print(f"Datos del problema:")
print(f"  Media poblacional bajo H0 (μ₀): {mu_0_exam_ht} puntos (Promedio Nacional)")
print(f"  Media muestral (x̄): {sample_mean_exam_ht} puntos (Promedio de la Clase)")
print(f"  Tamaño de muestra (n): {sample_size_exam_ht}")
print(f"  Desviación estándar poblacional (σ): {pop_std_dev_exam_ht} puntos")
print(f"  Nivel de significancia (α): {alpha_exam_ht}\n")


# 1. Plantear las hipótesis nula y alternativa
print("1. Planteamiento de Hipótesis:")
print(f"   Hipótesis Nula (H₀): μ = {mu_0_exam_ht} puntos (El promedio de la clase es igual al nacional)")
# La afirmación es que es "diferente", lo que implica una prueba bilateral.
print(f"   Hipótesis Alternativa (H₁): μ ≠ {mu_0_exam_ht} puntos (El promedio de la clase es diferente al nacional - Test bilateral)\n")


# 2. Calcule el estadístico Z.
# Fórmula para Test Z de una muestra con sigma conocida: Z = (x_bar - mu_0) / (sigma / sqrt(n))
standard_error_exam_ht = pop_std_dev_exam_ht / np.sqrt(sample_size_exam_ht)
z_statistic_exam_ht = (sample_mean_exam_ht - mu_0_exam_ht) / standard_error_exam_ht

print("2. Cálculo del Estadístico de Prueba Z:")
print(f"   Error estándar: σ / √n = {pop_std_dev_exam_ht} / √{sample_size_exam_ht} = {standard_error_exam_ht:.4f} puntos")
print(f"   Estadístico Z = (x̄ - μ₀) / Error estándar = ({sample_mean_exam_ht} - {mu_0_exam_ht}) / {standard_error_exam_ht:.4f} = {z_statistic_exam_ht:.4f}\n")


# 3. Determine la región crítica y la región de rechazo.
# Para un test bilateral con nivel de significancia alpha, el valor crítico Z es el valor
# que deja alpha/2 área en cada cola de la distribución normal estándar.
critical_value_lower_exam_ht = norm.ppf(alpha_exam_ht / 2)
critical_value_upper_exam_ht = norm.ppf(1 - alpha_exam_ht / 2)


print("3. Determinación del Valor Crítico y la Región de Rechazo:")
print(f"   Nivel de significancia (α): {alpha_exam_ht}")
print(f"   Para un test bilateral, los valores críticos Z son norm.ppf(α/2) y norm.ppf(1-α/2).")
print(f"   Valor crítico Z inferior: {critical_value_lower_exam_ht:.4f}")
print(f"   Valor crítico Z superior: {critical_value_upper_exam_ht:.4f}")
print(f"   Región de rechazo: Z < {critical_value_lower_exam_ht:.4f} o Z > {critical_value_upper_exam_ht:.4f}\n")


# Graficar la distribución normal estándar, región de rechazo, valor crítico y estadístico Z
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1) # Distribución Normal Estándar

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Normal Estándar (Z)')

# Sombrear la región de rechazo inferior
x_reject_lower_exam_ht = x[x < critical_value_lower_exam_ht]
y_reject_lower_exam_ht = norm.pdf(x_reject_lower_exam_ht, 0, 1)
plt.fill_between(x_reject_lower_exam_ht, 0, y_reject_lower_exam_ht, color='red', alpha=0.5, label='Región de Rechazo (α/2)')

# Sombrear la región de rechazo superior
x_reject_upper_exam_ht = x[x > critical_value_upper_exam_ht]
y_reject_upper_exam_ht = norm.pdf(x_reject_upper_exam_ht, 0, 1)
plt.fill_between(x_reject_upper_exam_ht, 0, y_reject_upper_exam_ht, color='red', alpha=0.5) # No se necesita etiqueta duplicada


# Agregar líneas verticales para los valores críticos y el estadístico de prueba
plt.axvline(critical_value_lower_exam_ht, color='black', linestyle='--', label=f'Valor Crítico Z Inferior ({critical_value_lower_exam_ht:.2f})')
plt.axvline(critical_value_upper_exam_ht, color='black', linestyle='--', label=f'Valor Crítico Z Superior ({critical_value_upper_exam_ht:.2f})')
plt.axvline(z_statistic_exam_ht, color='green', linestyle='-', label=f'Estadístico de Prueba Z ({z_statistic_exam_ht:.2f})', linewidth=2)


plt.title('Distribución Normal Estándar con Región de Rechazo y Estadístico de Prueba (Test Bilateral)')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# Calcular el valor p (para un test bilateral)
# El valor p es la probabilidad de obtener un estadístico Z tan extremo como el calculado, en cualquier cola.
p_value_exam_ht = 2 * norm.sf(abs(z_statistic_exam_ht))

print(f"Valor p calculado (bilateral): {p_value_exam_ht:.4f}\n")


# 4. Concluya si se rechaza H0.
print("4. Decisión Estadística y Conclusión:")

# Comparar el estadístico de prueba con el valor crítico O el valor p con alpha
print(f"Comparando el Estadístico Z ({z_statistic_exam_ht:.4f}) con los Valores Críticos Z ({critical_value_lower_exam_ht:.4f} y {critical_value_upper_exam_ht:.4f}):")
if z_statistic_exam_ht < critical_value_lower_exam_ht or z_statistic_exam_ht > critical_value_upper_exam_ht:
    print("   El Estadístico Z cae en la región de rechazo.")
    decision_exam_ht = "Rechazar H₀"
else:
    print("   El Estadístico Z NO cae en la región de rechazo.")
    decision_exam_ht = "No rechazar H₀"

print(f"\nComparando el Valor p ({p_value_exam_ht:.4f}) con α ({alpha_exam_ht}):")
if p_value_exam_ht <= alpha_exam_ht:
     print(f"   El Valor p ({p_value_exam_ht:.4f}) es menor o igual que α ({alpha_exam_ht}).")
     decision_p_value_exam_ht = "Rechazar H₀"
else:
     print(f"   El Valor p ({p_value_exam_ht:.4f}) es mayor que α ({alpha_exam_ht}).")
     decision_p_value_exam_ht = "No rechazar H₀"

print(f"\nDecisión Estadística: {decision_exam_ht}")

print("\nConclusión Práctica:")
if decision_exam_ht == "Rechazar H₀":
    print(f"   Al nivel de significancia del {alpha_exam_ht*100}%, hay suficiente evidencia para rechazar la hipótesis nula.")
    print(f"   Concluimos que el promedio de la clase ({sample_mean_exam_ht} puntos) es significativamente diferente del promedio nacional ({mu_0_exam_ht} puntos).")
else:
    print(f"   Al nivel de significancia del {alpha_exam_ht*100}%, no hay suficiente evidencia para rechazar la hipótesis nula.")
    print(f"   No podemos concluir que el promedio de la clase sea significativamente diferente del promedio nacional de {mu_0_exam_ht} puntos.")
    print("   La diferencia observada (78 puntos) podría deberse a la variación aleatoria de la muestra.")

print("\nInterpretación del Valor p:")
print(f"   El valor p ({p_value_exam_ht:.4f}) es la probabilidad de observar una media muestral tan extrema (78 o menos de 72 puntos),")
print(f"   si el verdadero promedio poblacional fuera 75 puntos (es decir, si H₀ fuera verdadera).")
print(f"   Como {p_value_exam_ht:.4f} > {alpha_exam_ht}, esta probabilidad es relativamente alta bajo H₀,")
print("   por lo que los datos no proporcionan evidencia fuerte contra H₀.")
```

```
from scipy.stats import t
import numpy as np
import matplotlib.pyplot as plt

# --- Prueba de Hipótesis Específica (Test t Unilateral Superior para la Media - Tiempo de Entrega) ---
print("\n--- Prueba de Hipótesis Específica (Test t Unilateral Superior para la Media - Tiempo de Entrega) ---\n")

# Datos del problema:
mu_0_delivery = 25      # Media poblacional bajo H0 (minutos)
sample_mean_delivery = 28   # Media de la muestra (minutos)
sample_size_delivery = 15    # Tamaño de la muestra
sample_std_dev_delivery = 4    # Desviación estándar de la MUESTRA (s) en minutos
alpha_delivery = 0.05      # Nivel de significancia

print(f"Datos del problema:")
print(f"  Media poblacional bajo H0 (μ₀): {mu_0_delivery} minutos (Tiempo afirmado)")
print(f"  Media muestral (x̄): {sample_mean_delivery} minutos (Tiempo observado)")
print(f"  Tamaño de muestra (n): {sample_size_delivery}")
print(f"  Desviación estándar de la muestra (s): {sample_std_dev_delivery} minutos")
print(f"  Nivel de significancia (α): {alpha_delivery}\n")


# 1. Plantear las hipótesis nula y alternativa
print("1. Planteamiento de Hipótesis:")
print(f"   Hipótesis Nula (H₀): μ = {mu_0_delivery} minutos (El tiempo promedio de entrega es 25 minutos)")
# La sospecha es que el tiempo es mayor.
print(f"   Hipótesis Alternativa (H₁): μ > {mu_0_delivery} minutos (El tiempo promedio de entrega es mayor - Test unilateral superior)\n")


# Como el tamaño de la muestra es pequeño (n < 30) Y la desviación estándar POBLACIONAL (σ) es desconocida
# (solo tenemos la desviación estándar de la MUESTRA 's'), debemos usar la distribución t de Student.
# Los grados de libertad (df) para un test t de una muestra son n - 1.
degrees_of_freedom_delivery = sample_size_delivery - 1

print(f"Usando la distribución t de Student con {degrees_of_freedom_delivery} grados de libertad.\n")


# 2. Calcular el estadístico t
# Fórmula para Test t de una muestra con sigma desconocida: t = (x_bar - mu_0) / (s / sqrt(n))
standard_error_delivery = sample_std_dev_delivery / np.sqrt(sample_size_delivery)
t_statistic_delivery = (sample_mean_delivery - mu_0_delivery) / standard_error_delivery

print("2. Cálculo del Estadístico de Prueba t:")
print(f"   Error estándar muestral: s / √n = {sample_std_dev_delivery} / √{sample_size_delivery} = {standard_error_delivery:.4f} minutos")
print(f"   Estadístico t = (x̄ - μ₀) / Error estándar muestral = ({sample_mean_delivery} - {mu_0_delivery}) / {standard_error_delivery:.4f} = {t_statistic_delivery:.4f}\n")


# 3. Determine la región crítica y la región de rechazo.
# Para un test unilateral superior con nivel de significancia alpha y df grados de libertad,
# el valor crítico t es el valor que deja 'alpha' área en la cola superior de la distribución t.
critical_value_delivery = t.ppf(1 - alpha_delivery, df=degrees_of_freedom_delivery)


print("3. Determinación del Valor Crítico y la Región de Rechazo:")
print(f"   Nivel de significancia (α): {alpha_delivery}")
print(f"   Grados de libertad (df): {degrees_of_freedom_delivery}")
print(f"   Para un test t unilateral superior, el valor crítico t es t.ppf(1 - α, df).")
print(f"   Valor crítico t: {critical_value_delivery:.4f}")
print(f"   Región de rechazo: t > {critical_value_delivery:.4f}\n")


# Graficar la distribución t, región de rechazo, valor crítico y estadístico t
# Usamos la distribución t con los grados de libertad calculados
x = np.linspace(t.ppf(0.0001, df=degrees_of_freedom_delivery), t.ppf(0.9999, df=degrees_of_freedom_delivery), 1000)
y = t.pdf(x, df=degrees_of_freedom_delivery) # Distribución t de Student


plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Distribución t (df={degrees_of_freedom_delivery})')

# Sombrear la región de rechazo (cola superior)
x_reject_delivery = x[x > critical_value_delivery]
y_reject_delivery = t.pdf(x_reject_delivery, df=degrees_of_freedom_delivery)
plt.fill_between(x_reject_delivery, 0, y_reject_delivery, color='red', alpha=0.5, label='Región de Rechazo (α)')


# Agregar líneas verticales para el valor crítico y el estadístico de prueba
plt.axvline(critical_value_delivery, color='black', linestyle='--', label=f'Valor Crítico t ({critical_value_delivery:.2f})')
plt.axvline(t_statistic_delivery, color='green', linestyle='-', label=f'Estadístico de Prueba t ({t_statistic_delivery:.2f})', linewidth=2)


plt.title(f'Distribución t de Student (df={degrees_of_freedom_delivery}) con Región de Rechazo y Estadístico de Prueba (Test Unilateral Superior)')
plt.xlabel('Valor t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# Calcular el valor p (para un test t unilateral superior)
# El valor p es la probabilidad de obtener un estadístico t tan grande o más grande que el observado.
p_value_delivery = t.sf(t_statistic_delivery, df=degrees_of_freedom_delivery)

print(f"Valor p calculado (unilateral superior): {p_value_delivery:.4f}\n")


# 4. Concluya si se rechaza H0.
print("4. Decisión Estadística y Conclusión:")

# Comparar el estadístico de prueba con el valor crítico O el valor p con alpha
print(f"Comparando el Estadístico t ({t_statistic_delivery:.4f}) con el Valor Crítico t ({critical_value_delivery:.4f}):")
if t_statistic_delivery > critical_value_delivery:
    print("   El Estadístico t cae en la región de rechazo.")
    decision_delivery = "Rechazar H₀"
else:
    print("   El Estadístico t NO cae en la región de rechazo.")
    decision_delivery = "No rechazar H₀"

print(f"\nComparando el Valor p ({p_value_delivery:.4f}) con α ({alpha_delivery}):")
if p_value_delivery <= alpha_delivery:
     print(f"   El Valor p ({p_value_delivery:.4f}) es menor o igual que α ({alpha_delivery}).")
     decision_p_value_delivery = "Rechazar H₀"
else:
     print(f"   El Valor p ({p_value_delivery:.4f}) es mayor que α ({alpha_delivery}).")
     decision_p_value_delivery = "No rechazar H₀"

print(f"\nDecisión Estadística: {decision_delivery}")

print("\nConclusión Práctica:")
if decision_delivery == "Rechazar H₀":
    print(f"   Al nivel de significancia del {alpha_delivery*100}%, hay suficiente evidencia para rechazar la hipótesis nula.")
    print(f"   Concluimos que el tiempo promedio de entrega a domicilio es significativamente mayor a {mu_0_delivery} minutos.")
else:
    print(f"   Al nivel de significancia del {alpha_delivery*100}%, no hay suficiente evidencia para rechazar la hipótesis nula.")
    print(f"   No podemos concluir que el tiempo promedio de entrega a domicilio sea significativamente mayor a {mu_0_delivery} minutos.")
    print("   La diferencia observada de 28 minutos podría deberse a la variación aleatoria de la muestra.")

print("\nInterpretación del Valor p:")
print(f"   El valor p ({p_value_delivery:.4f}) es la probabilidad de observar una media muestral de 28 minutos o más,")
print(f"   si el verdadero tiempo promedio poblacional fuera 25 minutos (es decir, si H₀ fuera verdadera).")
print(f"   Como {p_value_delivery:.4f} > {alpha_delivery}, esta probabilidad es relativamente alta bajo H₀,")
print("   por lo que los datos no proporcionan evidencia fuerte contra H₀.")
```

```
from scipy.stats import t
import numpy as np
import matplotlib.pyplot as plt

# --- Prueba de Hipótesis Específica (Test t Unilateral Inferior para la Media - Focos) ---
print("\n--- Prueba de Hipótesis Específica (Test t Unilateral Inferior para la Media - Focos) ---\n")

# Datos del problema:
mu_0_bulbs = 10000      # Media poblacional bajo H0 (horas)
sample_mean_bulbs = 9800   # Media de la muestra (horas)
sample_size_bulbs = 20    # Tamaño de la muestra
sample_std_dev_bulbs = 300    # Desviación estándar de la MUESTRA (s) en horas
alpha_bulbs = 0.05      # Nivel de significancia

print(f"Datos del problema:")
print(f"  Media poblacional bajo H0 (μ₀): {mu_0_bulbs} horas (Duración afirmada)")
print(f"  Media muestral (x̄): {sample_mean_bulbs} horas (Duración observada)")
print(f"  Tamaño de muestra (n): {sample_size_bulbs}")
print(f"  Desviación estándar de la muestra (s): {sample_std_dev_bulbs} horas")
print(f"  Nivel de significancia (α): {alpha_bulbs}\n")


# 1. Plantear las hipótesis nula y alternativa
print("1. Planteamiento de Hipótesis:")
print(f"   Hipótesis Nula (H₀): μ = {mu_0_bulbs} horas (La duración promedio es 10,000 horas)")
# La muestra es menor, lo que sugiere que la duración puede ser menor.
print(f"   Hipótesis Alternativa (H₁): μ < {mu_0_bulbs} horas (La duración promedio es menor - Test unilateral inferior)\n")


# Como el tamaño de la muestra es pequeño (n < 30) Y la desviación estándar POBLACIONAL (σ) es desconocida
# (solo tenemos la desviación estándar de la MUESTRA 's'), debemos usar la distribución t de Student.
# Los grados de libertad (df) para un test t de una muestra son n - 1.
degrees_of_freedom_bulbs = sample_size_bulbs - 1

print(f"Usando la distribución t de Student con {degrees_of_freedom_bulbs} grados de libertad.\n")


# 2. Calcular el estadístico t
# Fórmula para Test t de una muestra con sigma desconocida: t = (x_bar - mu_0) / (s / sqrt(n))
standard_error_bulbs = sample_std_dev_bulbs / np.sqrt(sample_size_bulbs)
t_statistic_bulbs = (sample_mean_bulbs - mu_0_bulbs) / standard_error_bulbs

print("2. Cálculo del Estadístico de Prueba t:")
print(f"   Error estándar muestral: s / √n = {sample_std_dev_bulbs} / √{sample_size_bulbs} = {standard_error_bulbs:.4f} horas")
print(f"   Estadístico t = (x̄ - μ₀) / Error estándar muestral = ({sample_mean_bulbs} - {mu_0_bulbs}) / {standard_error_bulbs:.4f} = {t_statistic_bulbs:.4f}\n")


# 3. Determine la región crítica y la región de rechazo.
# Para un test unilateral inferior con nivel de significancia alpha y df grados de libertad,
# el valor crítico t es el valor que deja 'alpha' área en la cola inferior de la distribución t.
critical_value_bulbs = t.ppf(alpha_bulbs, df=degrees_of_freedom_bulbs)


print("3. Determinación del Valor Crítico y la Región de Rechazo:")
print(f"   Nivel de significancia (α): {alpha_bulbs}")
print(f"   Grados de libertad (df): {degrees_of_freedom_bulbs}")
print(f"   Para un test t unilateral inferior, el valor crítico t es t.ppf(α, df).")
print(f"   Valor crítico t: {critical_value_bulbs:.4f}")
print(f"   Región de rechazo: t < {critical_value_bulbs:.4f}\n")


# Graficar la distribución t, región de rechazo, valor crítico y estadístico t
# Usamos la distribución t con los grados de libertad calculados
x = np.linspace(t.ppf(0.0001, df=degrees_of_freedom_bulbs), t.ppf(0.9999, df=degrees_of_freedom_bulbs), 1000)
y = t.pdf(x, df=degrees_of_freedom_bulbs) # Distribución t de Student


plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Distribución t (df={degrees_of_freedom_bulbs})')

# Sombrear la región de rechazo (cola inferior)
x_reject_bulbs = x[x < critical_value_bulbs]
y_reject_bulbs = t.pdf(x_reject_bulbs, df=degrees_of_freedom_bulbs)
plt.fill_between(x_reject_bulbs, 0, y_reject_bulbs, color='red', alpha=0.5, label='Región de Rechazo (α)')


# Agregar líneas verticales para el valor crítico y el estadístico de prueba
plt.axvline(critical_value_bulbs, color='black', linestyle='--', label=f'Valor Crítico t ({critical_value_bulbs:.2f})')
plt.axvline(t_statistic_bulbs, color='green', linestyle='-', label=f'Estadístico de Prueba t ({t_statistic_bulbs:.2f})', linewidth=2)


plt.title(f'Distribución t de Student (df={degrees_of_freedom_bulbs}) con Región de Rechazo y Estadístico de Prueba (Test Unilateral Inferior)')
plt.xlabel('Valor t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# Calcular el valor p (para un test t unilateral inferior)
# El valor p es la probabilidad de obtener un estadístico t tan pequeño o más pequeño que el observado.
p_value_bulbs = t.cdf(t_statistic_bulbs, df=degrees_of_freedom_bulbs)

print(f"Valor p calculado (unilateral inferior): {p_value_bulbs:.4f}\n")


# 4. Concluya si se rechaza H0.
print("4. Decisión Estadística y Conclusión:")

# Comparar el estadístico de prueba con el valor crítico O el valor p con alpha
print(f"Comparando el Estadístico t ({t_statistic_bulbs:.4f}) con el Valor Crítico t ({critical_value_bulbs:.4f}):")
if t_statistic_bulbs < critical_value_bulbs:
    print("   El Estadístico t cae en la región de rechazo.")
    decision_bulbs = "Rechazar H₀"
else:
    print("   El Estadístico t NO cae en la región de rechazo.")
    decision_bulbs = "No rechazar H₀"

print(f"\nComparando el Valor p ({p_value_bulbs:.4f}) con α ({alpha_bulbs}):")
if p_value_bulbs <= alpha_bulbs:
     print(f"   El Valor p ({p_value_bulbs:.4f}) es menor o igual que α ({alpha_bulbs}).")
     decision_p_value_bulbs = "Rechazar H₀"
else:
     print(f"   El Valor p ({p_value_bulbs:.4f}) es mayor que α ({alpha_bulbs}).")
     decision_p_value_bulbs = "No rechazar H₀"

print(f"\nDecisión Estadística: {decision_bulbs}")

print("\nConclusión Práctica:")
if decision_bulbs == "Rechazar H₀":
    print(f"   Al nivel de significancia del {alpha_bulbs*100}%, hay suficiente evidencia para rechazar la hipótesis nula.")
    print(f"   Concluimos que la duración promedio de los focos es significativamente menor a {mu_0_bulbs} horas.")
else:
    print(f"   Al nivel de significancia del {alpha_bulbs*100}%, no hay suficiente evidencia para rechazar la hipótesis nula.")
    print(f"   No podemos concluir que la duración promedio de los focos sea significativamente menor a {mu_0_bulbs} horas.")
    print("   La diferencia observada de 9800 horas podría deberse a la variación aleatoria de la muestra.")

print("\nInterpretación del Valor p:")
print(f"   El valor p ({p_value_bulbs:.4f}) es la probabilidad de obtener un estadístico t tan pequeño o más pequeño que el observado,")
print(f"   si la verdadera media poblacional fuera {mu_0_bulbs} horas (es decir, si H₀ fuera verdadera).")
print(f"   Como {p_value_bulbs:.4f} > {alpha_bulbs}, esta probabilidad es relativamente alta bajo H₀,")
print("   por lo que los datos no proporcionan evidencia fuerte contra H₀.")
```

```
from scipy.stats import chi2
import numpy as np
import matplotlib.pyplot as plt

# --- Prueba de Hipótesis Específica (Test Chi-cuadrado para la Varianza - Tiempo de Llamada) ---
print("\n--- Prueba de Hipótesis Específica (Test Chi-cuadrado para la Varianza - Tiempo de Llamada) ---\n")

# Datos del problema:
sigma2_0_calls = 2.5      # Varianza poblacional bajo H0 (minutos^2)
sample_variance_calls = 4   # Varianza de la muestra (minutos^2)
sample_size_calls = 16    # Tamaño de la muestra
alpha_calls = 0.05      # Nivel de significancia

print(f"Datos del problema:")
print(f"  Varianza poblacional bajo H0 (σ²₀): {sigma2_0_calls} minutos²")
print(f"  Varianza muestral (s²): {sample_variance_calls} minutos²")
print(f"  Tamaño de muestra (n): {sample_size_calls}")
print(f"  Nivel de significancia (α): {alpha_calls}\n")


# 1. Plantear las hipótesis nula y alternativa
print("1. Planteamiento de Hipótesis:")
print(f"   Hipótesis Nula (H₀): σ² = {sigma2_0_calls} minutos² (La varianza poblacional es 2.5 minutos²)")
# La afirmación es que "difiere", lo que implica una prueba bilateral.
print(f"   Hipótesis Alternativa (H₁): σ² ≠ {sigma2_0_calls} minutos² (La varianza poblacional difiere de 2.5 minutos² - Test bilateral)\n")


# 2. Calcule el estadístico Chi-cuadrado.
# Fórmula para Test Chi-cuadrado de una muestra para la varianza: χ² = ((n - 1) * s²) / σ²₀
degrees_of_freedom_calls = sample_size_calls - 1
chi2_statistic_calls = ((sample_size_calls - 1) * sample_variance_calls) / sigma2_0_calls

print("2. Cálculo del Estadístico de Prueba Chi-cuadrado:")
print(f"   Grados de libertad (df): n - 1 = {sample_size_calls} - 1 = {degrees_of_freedom_calls}")
print(f"   Estadístico χ² = ((n - 1) * s²) / σ²₀ = (({sample_size_calls} - 1) * {sample_variance_calls}) / {sigma2_0_calls} = {chi2_statistic_calls:.4f}\n")


# 3. Determine los grados de libertad y los valores críticos, además de la región crítica y la región de rechazo.
# Para un test bilateral con nivel de significancia alpha y df grados de libertad,
# el área alpha se divide en las dos colas de la distribución Chi-cuadrado.
# El valor crítico inferior deja alpha/2 área en la cola inferior.
critical_value_lower_calls = chi2.ppf(alpha_calls / 2, df=degrees_of_freedom_calls)
# El valor crítico superior deja alpha/2 área en la cola superior (o 1 - alpha/2 área a la izquierda).
critical_value_upper_calls = chi2.ppf(1 - alpha_calls / 2, df=degrees_of_freedom_calls)


print("3. Determinación de los Valores Críticos y la Región de Rechazo:")
print(f"   Nivel de significancia (α): {alpha_calls}")
print(f"   Grados de libertad (df): {degrees_of_freedom_calls}")
print(f"   Para un test bilateral, los valores críticos χ² son chi2.ppf(α/2, df) y chi2.ppf(1-α/2, df).")
print(f"   Valor crítico χ² inferior: {critical_value_lower_calls:.4f}")
print(f"   Valor crítico χ² superior: {critical_value_upper_calls:.4f}")
print(f"   Región de rechazo: χ² < {critical_value_lower_calls:.4f} o χ² > {critical_value_upper_calls:.4f}\n")


# Graficar la distribución Chi-cuadrado, región de rechazo, valores críticos y estadístico Chi-cuadrado
# El rango del eje x debe comenzar en 0 para la distribución Chi-cuadrado.
# Usamos chi2.ppf para encontrar un rango apropiado basado en los percentiles.
x = np.linspace(chi2.ppf(0.0001, df=degrees_of_freedom_calls), chi2.ppf(0.9999, df=degrees_of_freedom_calls), 1000)
y = chi2.pdf(x, df=degrees_of_freedom_calls) # Distribución Chi-cuadrado

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Distribución χ² (df={degrees_of_freedom_calls})')

# Sombrear la región de rechazo inferior
x_reject_lower_calls = x[x < critical_value_lower_calls]
y_reject_lower_calls = chi2.pdf(x_reject_lower_calls, df=degrees_of_freedom_calls)
plt.fill_between(x_reject_lower_calls, 0, y_reject_lower_calls, color='red', alpha=0.5, label='Región de Rechazo (α/2)')

# Sombrear la región de rechazo superior
x_reject_upper_calls = x[x > critical_value_upper_calls]
y_reject_upper_calls = chi2.pdf(x_reject_upper_calls, df=degrees_of_freedom_calls)
plt.fill_between(x_reject_upper_calls, 0, y_reject_upper_calls, color='red', alpha=0.5) # No se necesita etiqueta duplicada


# Agregar líneas verticales para los valores críticos y el estadístico de prueba
plt.axvline(critical_value_lower_calls, color='black', linestyle='--', label=f'Valor Crítico χ² Inferior ({critical_value_lower_calls:.2f})')
plt.axvline(critical_value_upper_calls, color='black', linestyle='--', label=f'Valor Crítico χ² Superior ({critical_value_upper_calls:.2f})')
plt.axvline(chi2_statistic_calls, color='green', linestyle='-', label=f'Estadístico de Prueba χ² ({chi2_statistic_calls:.2f})', linewidth=2)


plt.title(f'Distribución χ² (df={degrees_of_freedom_calls}) con Región de Rechazo y Estadístico de Prueba (Test Bilateral)')
plt.xlabel('Valor χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# Calcular el valor p (para un test bilateral)
# Para un test bilateral, el valor p es 2 * el área en la cola más extrema.
# Si el estadístico está a la izquierda del centro (df), calculamos 2 * CDF(estadístico).
# Si el estadístico está a la derecha del centro, calculamos 2 * SF(estadístico).
# El centro de la distribución Chi-cuadrado es igual a los grados de libertad (df).
if chi2_statistic_calls < degrees_of_freedom_calls:
    p_value_calls = 2 * chi2.cdf(chi2_statistic_calls, df=degrees_of_freedom_calls)
    print(f"Estadístico χ² ({chi2_statistic_calls:.4f}) es menor que df ({degrees_of_freedom_calls}). Valor p = 2 * CDF({chi2_statistic_calls:.4f}).")
else:
    p_value_calls = 2 * chi2.sf(chi2_statistic_calls, df=degrees_of_freedom_calls)
    print(f"Estadístico χ² ({chi2_statistic_calls:.4f}) es mayor o igual que df ({degrees_of_freedom_calls}). Valor p = 2 * SF({chi2_statistic_calls:.4f}).")

print(f"Valor p calculado (bilateral): {p_value_calls:.4f}\n")


# 4. Concluya si se rechaza H0.
print("4. Decisión Estadística y Conclusión:")

# Comparar el estadístico de prueba con el valor crítico O el valor p con alpha
print(f"Comparando el Estadístico χ² ({chi2_statistic_calls:.4f}) con los Valores Críticos χ² ({critical_value_lower_calls:.4f} y {critical_value_upper_calls:.4f}):")
if chi2_statistic_calls < critical_value_lower_calls or chi2_statistic_calls > critical_value_upper_calls:
    print("   El Estadístico χ² cae en la región de rechazo.")
    decision_calls = "Rechazar H₀"
else:
    print("   El Estadístico χ² NO cae en la región de rechazo.")
    decision_calls = "No rechazar H₀"

print(f"\nComparando el Valor p ({p_value_calls:.4f}) con α ({alpha_calls}):")
if p_value_calls <= alpha_calls:
     print(f"   El Valor p ({p_value_calls:.4f}) es menor o igual que α ({alpha_calls}).")
     decision_p_value_calls = "Rechazar H₀"
else:
     print(f"   El Valor p ({p_value_calls:.4f}) es mayor que α ({alpha_calls}).")
     decision_p_value_calls = "No rechazar H₀"

print(f"\nDecisión Estadística: {decision_calls}")

print("\nConclusión Práctica:")
if decision_calls == "Rechazar H₀":
    print(f"   Al nivel de significancia del {alpha_calls*100}%, hay suficiente evidencia para rechazar la hipótesis nula.")
    print(f"   Concluimos que la varianza poblacional del tiempo de duración de las llamadas es significativamente diferente de {sigma2_0_calls} minutos².")
else:
    print(f"   Al nivel de significancia del {alpha_calls*100}%, no hay suficiente evidencia para rechazar la hipótesis nula.")
    print(f"   No podemos concluir que la varianza poblacional del tiempo de duración de las llamadas sea significativamente diferente de {sigma2_0_calls} minutos².")
    print(f"   La diferencia observada en la varianza muestral ({sample_variance_calls} minutos²) podría deberse a la variación aleatoria de la muestra.")

print("\nInterpretación del Valor p:")
print(f"   El valor p ({p_value_calls:.4f}) es la probabilidad de observar una varianza muestral tan extrema (4 minutos² o más, o un valor simétricamente extremo en la cola inferior),")
print(f"   si la verdadera varianza poblacional fuera {sigma2_0_calls} minutos² (es decir, si H₀ fuera verdadera).")
print(f"   Como {p_value_calls:.4f} > {alpha_calls}, esta probabilidad es relativamente alta bajo H₀,")
print("   por lo que los datos no proporcionan evidencia fuerte contra H₀.")
```

```
# -*- coding: utf-8 -*-
"""Cuaderno de Google Colab: Pruebas de Hipótesis, Intervalos de Confianza y Tamaño de Muestra

Este cuaderno interactivo explora y demuestra los conceptos fundamentales
de las pruebas de hipótesis, los intervalos de confianza y la determinación
del tamaño de muestra en estadística, utilizando Python y librerías comunes
como NumPy, SciPy y Matplotlib.

Contenido:
1.  Introducción a los conceptos.
2.  Pruebas de Hipótesis: pasos, cálculo e interpretación.
3.  Intervalos de Confianza: cálculo e interpretación.
4.  Tamaño de Muestra: cálculo para estimación y pruebas de hipótesis.
"""

# @title 1. Instalación e Importación de Librerías
# @markdown Ejecuta esta celda para instalar e importar las librerías necesarias.

# Instalación de librerías (si no están preinstaladas en Colab)
# !pip install numpy scipy matplotlib statsmodels

# Importación de librerías
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm # Usado para distribuciones Normales (Z)
import math # Usado para funciones matemáticas como ceil
try:
    from statsmodels.stats.power import NormalIndPower # Usado para análisis de poder y tamaño de muestra con statsmodels
    _STATMODELS_AVAILABLE = True
except ImportError:
    _STATMODELS_AVAILABLE = False
    print("Advertencia: La librería statsmodels no está instalada. Algunas secciones de cálculo de tamaño de muestra no funcionarán.")


# Configuración para matplotlib (inline en cuadernos)
# %matplotlib inline

print("Librerías instaladas e importadas.")

# @title 2. Introducción a los Conceptos Estadísticos
# @markdown Lee esta sección para entender la base de las pruebas de hipótesis, intervalos de confianza y tamaño de muestra.

print("""# Introducción a las Pruebas de Hipótesis, Intervalos de Confianza y Tamaño de Muestra

En el análisis de datos y la inferencia estadística, utilizamos herramientas para tomar decisiones informadas sobre una población basándonos en una muestra.

*   **Pruebas de Hipótesis:** Un método formal para determinar si hay suficiente evidencia en una muestra de datos para rechazar una hipótesis nula sobre una población. Nos ayudan a evaluar si una observación es estadísticamente significativa o si podría deberse simplemente al azar.

*   **Intervalos de Confianza:** Proporcionan un rango de valores dentro del cual es probable que se encuentre un parámetro de población desconocido. Nos dan una idea de la precisión de nuestra estimación basada en la muestra.

*   **Tamaño de la Muestra:** Se refiere al número de observaciones o individuos incluidos en un estudio. Determinar un tamaño de muestra adecuado es crucial porque un tamaño de muestra insuficiente puede llevar a conclusiones poco fiables, mientras que uno excesivamente grande puede ser ineficiente en términos de recursos.

Estos conceptos son fundamentales en la práctica porque nos permiten sacar conclusiones fiables de los datos, cuantificar la incertidumbre de nuestras estimaciones y diseñar estudios eficientes que maximicen la probabilidad de detectar efectos reales si existen.
""")


# @title 3. Pruebas de Hipótesis: Pasos y Demostración
# @markdown Explora los 7 pasos de una prueba de hipótesis y visualiza los conceptos clave como la región de rechazo y el valor p.

print("""## Pruebas de Hipótesis: Pasos y Demostración

Una prueba de hipótesis sigue una serie de pasos para evaluar una afirmación sobre un parámetro poblacional. Aquí detallamos los pasos y usamos un ejemplo para ilustrarlos.

### Pasos de una Prueba de Hipótesis:

**Paso 1: Definir las Hipótesis**
*   **Hipótesis Nula (H0):** La afirmación que se asume como verdadera hasta que haya evidencia suficiente en su contra (ej. no hay efecto, no hay diferencia).
*   **Hipótesis Alternativa (H1):** La afirmación que se desea probar; lo que se concluye si se rechaza H0 (ej. hay un efecto, hay una diferencia).

**Paso 2: Establecer el Nivel de Significancia (Alpha - α)**
*   La probabilidad máxima permitida de cometer un Error de Tipo I (rechazar H0 cuando es verdadera). Comúnmente 0.05.

**Paso 3: Identificar el Estadístico de Prueba**
*   Un valor calculado a partir de la muestra que mide cuánto se desvían los datos de lo esperado bajo H0 (ej. estadístico Z, t, F).

**Paso 4: Determinar el Valor Crítico y la Región de Rechazo**
*   **Valor Crítico:** El umbral en la distribución del estadístico de prueba que separa la región de no rechazo de la región de rechazo.
*   **Región de Rechazo:** El conjunto de valores del estadístico de prueba que llevan a rechazar H0.

**Paso 5: Calcular el Estadístico de Prueba y el Valor p**
*   Calcular el estadístico de prueba usando los datos de la muestra.
*   **Valor p:** La probabilidad de obtener un estadístico de prueba tan extremo o más extremo que el observado, asumiendo que H0 es verdadera.

**Paso 6: Tomar una Decisión Estadística**
*   Comparar el estadístico de prueba con el valor crítico, o el valor p con alpha. Si el estadístico de prueba cae en la región de rechazo (o p-value <= alpha), se rechaza H0.

**Paso 7: Extraer una Conclusión Práctica**
*   Interpretar la decisión estadística en el contexto del problema. Indicar si hay suficiente evidencia para apoyar H1 o no hay suficiente evidencia para rechazar H0.

### Demostración con Código (Test Z Bilateral para la Media)

Consideremos un ejemplo hipotético: queremos probar si la media de una población es diferente de 50, con un nivel de significancia del 5%. Asumimos que conocemos la desviación estándar poblacional (σ = 10).
""")

# @title Paso 1-4: Hipótesis, Alpha, Valor Crítico y Región de Rechazo
# @markdown Define las hipótesis, el nivel de significancia y calcula/visualiza los valores críticos y la región de rechazo para un test Z bilateral.

# Paso 1: Definir las Hipótesis
# H0: μ = 50
# H1: μ ≠ 50 (Test Bilateral)

# Paso 2: Establecer el Nivel de Significancia (Alpha)
alpha = 0.05
print(f"Nivel de significancia (alpha): {alpha}")

# Paso 4: Determinar Valores Críticos y Región de Rechazo para un test bilateral Z
# Para un test Z bilateral, alpha se divide en las dos colas de la distribución.
# Usamos norm.ppf (inverse CDF) para encontrar el valor Z que deja alpha/2 en cada cola.
critical_value_lower = norm.ppf(alpha / 2)
critical_value_upper = norm.ppf(1 - alpha / 2)

print(f"Valor crítico inferior (Z) para alpha/2: {critical_value_lower:.4f}")
print(f"Valor crítico superior (Z) para 1-alpha/2: {critical_value_upper:.4f}")
print(f"Región de rechazo: Z < {critical_value_lower:.4f} o Z > {critical_value_upper:.4f}\n")


# Graficar la distribución normal estándar y sombrear la región de rechazo (Visualización Paso 4)
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1) # Distribución Normal Estándar (media=0, desv. estándar=1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Normal Estándar (Z)')

# Sombrear la región de rechazo inferior
x_reject_lower = x[x < critical_value_lower]
y_reject_lower = norm.pdf(x_reject_lower, 0, 1)
plt.fill_between(x_reject_lower, 0, y_reject_lower, color='red', alpha=0.5, label='Región de Rechazo (α/2)')

# Sombrear la región de rechazo superior
x_reject_upper = x[x > critical_value_upper]
y_reject_upper = norm.pdf(x_reject_upper, 0, 1)
plt.fill_between(x_reject_upper, 0, y_reject_upper, color='red', alpha=0.5) # No se necesita etiqueta duplicada


plt.title('Distribución Normal Estándar con Región de Rechazo (Test Bilateral)')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.axvline(critical_value_lower, color='black', linestyle='--', label=f'Valor Crítico Inferior ({critical_value_lower:.2f})')
plt.axvline(critical_value_upper, color='black', linestyle='--', label=f'Valor Crítico Superior ({critical_value_upper:.2f})')
plt.legend()
plt.grid(True)
plt.show()

print("""La gráfica muestra la distribución normal estándar. Las áreas sombreadas en rojo representan la región de rechazo para un test bilateral con alpha = 0.05.
Si el estadístico de prueba Z calculado a partir de los datos de la muestra cae dentro de estas áreas sombreadas,
rechazaríamos la hipótesis nula (H0) al nivel de significancia de 5%.
""")

# @title Paso 5-6: Calcular Estadístico de Prueba, Valor p y Graficar Resultados
# @markdown Simula datos de muestra, calcula el estadístico de prueba y el valor p, y grafica su posición relative a la región de rechazo.

# Simular datos de muestra para nuestro escenario hipotético
n = 30      # Tamaño de la muestra
x_bar = 53  # Media de la muestra observada
mu_0 = 50   # Media poblacional bajo H0 (de H0: μ = 50)
sigma = 10  # Desviación estándar poblacional (supuesta conocida)

print(f"Escenario simulado: Tamaño de muestra n = {n}, Media de la muestra x̄ = {x_bar}")
print(f"Hipótesis Nula H0: μ = {mu_0}, Desviación estándar de la población σ = {sigma}\n")

# Paso 5: Calcular el Estadístico de Prueba Z
# Fórmula para Test Z de una muestra con sigma conocida: Z = (x_bar - mu_0) / (sigma / sqrt(n))
z_statistic = (x_bar - mu_0) / (sigma / np.sqrt(n))

# Calcular el valor p (para un test bilateral)
# El valor p es la probabilidad de observar un estadístico Z tan extremo como el calculado, en cualquier cola.
# Usamos norm.sf (función de supervivencia = 1 - CDF) para la cola superior y multiplicamos por 2 para bilateral.
p_value = 2 * norm.sf(abs(z_statistic))

# Imprimir el estadístico de prueba calculado y el valor p
print(f"Estadístico de Prueba Z calculado: {z_statistic:.4f}")
print(f"Valor p calculado (bilateral): {p_value:.4f}\n")


# Paso 6: Graficar los resultados (distribución, valores críticos, estadístico de prueba)
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1) # Distribución Normal Estándar

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Normal Estándar (Z)')

# Sombrear la región de rechazo (usando los valores críticos calculados previamente)
x_reject_lower = x[x < critical_value_lower]
y_reject_lower = norm.pdf(x_reject_lower, 0, 1)
plt.fill_between(x_reject_lower, 0, y_reject_lower, color='red', alpha=0.5, label='Región de Rechazo (α/2)')

x_reject_upper = x[x > critical_value_upper]
y_reject_upper = norm.pdf(x_reject_upper, 0, 1)
plt.fill_between(x_reject_upper, 0, y_reject_upper, color='red', alpha=0.5)


# Agregar líneas verticales para los valores críticos y el estadístico de prueba
plt.axvline(critical_value_lower, color='black', linestyle='--', label=f'Valor Crítico Inferior ({critical_value_lower:.2f})')
plt.axvline(critical_value_upper, color='black', linestyle='--', label=f'Valor Crítico Superior ({critical_value_upper:.2f})')
plt.axvline(z_statistic, color='green', linestyle='-', label=f'Estadístico de Prueba Z ({z_statistic:.2f})', linewidth=2)


plt.title('Distribución Normal Estándar con Región de Rechazo, Valores Críticos y Estadístico de Prueba')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

print("""La gráfica visualiza el estadístico de prueba Z calculado (línea verde) en relación con los valores críticos (líneas negras) y la región de rechazo (áreas rojas).
Si la línea verde cae en las áreas rojas, rechazamos la hipótesis nula.
""")

# @title Paso 7: Decisión Estadística y Conclusión Práctica
# @markdown Interpreta los resultados de la prueba de hipótesis basándose en el estadístico de prueba, el valor p y el nivel de significancia.

print("""## Paso 7: Decisión Estadística y Conclusión Práctica

Una vez calculados el estadístico de prueba y el valor p, tomamos una decisión estadística y la traducimos en una conclusión práctica.

### Decisión Estadística:

*   **Usando el Valor Crítico:** Rechazamos H0 si el estadístico de prueba cae en la región de rechazo (Z <= critical_value_lower o Z >= critical_value_upper para un test bilateral).
*   **Usando el Valor p:** Rechazamos H0 si el valor p es menor o igual que alpha (p-value <= alpha).

### Conclusión Práctica:

*   Traduce la decisión estadística al contexto del problema. Evita decir "aceptar H0"; si no se rechaza H0, simplemente significa que no hay suficiente evidencia en la muestra para rechazarla al nivel de significancia elegido.

### Significado e Interpretación del Valor p:

*   El valor p es la probabilidad de observar los datos (o resultados más extremos) si H0 fuera verdadera.
*   Un valor p pequeño (< alpha) sugiere que los datos son improbables bajo H0, proporcionando evidencia contra H0.
*   Un valor p grande (> alpha) sugiere que los datos son razonablemente probables bajo H0, por lo que no hay evidencia suficiente para rechazar H0.

### Ejemplos de Interpretación (basado en nuestro ejemplo):
""")

print(f"Nuestro estadístico de prueba Z calculado fue {z_statistic:.4f}.")
print(f"Nuestro valor p calculado fue {p_value:.4f}.")
print(f"Nuestro nivel de significancia alpha fue {alpha}.\n")

# Escenario 1: Basado en nuestros resultados actuales (no rechazar H0)
print("#### Escenario de nuestro Ejemplo:")
print(f"**Decisión Estadística (usando el valor p):** Como el valor p ({p_value:.4f}) es > alpha ({alpha}), no rechazamos la hipótesis nula.")
print(f"**Decisión Estadística (usando el valor crítico):** Como el estadístico de prueba Z ({z_statistic:.4f}) está entre los valores críticos ({critical_value_lower:.4f} y {critical_value_upper:.4f}), no rechazamos la hipótesis nula.")
print("""**Conclusión Práctica:** No hay suficiente evidencia, al nivel de significancia de 5%, para concluir que la media de la población es significativamente diferente de 50. Esto no significa que la media sea exactamente 50, solo que la evidencia de esta muestra no es lo suficientemente fuerte para afirmar que es diferente.

""")

# Escenario 2: Caso hipotético donde H0 es rechazado
hypothetical_z_reject = 2.5 # Supongamos un Estadístico Z de 2.5
hypothetical_p_reject = 2 * norm.sf(abs(hypothetical_z_reject)) # Calcular el p-value para este Z hipotético

print("#### Escenario Hipotético (donde se rechazaría H0):")
print(f"Supongamos que el estadístico de prueba Z fuera {hypothetical_z_reject:.4f} y el valor p fuera {hypothetical_p_reject:.4f}.")
print(f"**Decisión Estadística (usando el valor p):** Como el valor p ({hypothetical_p_reject:.4f}) es <= alpha ({alpha}), rechazamos la hipótesis nula.")
print(f"**Decisión Estadística (usando el valor crítico):** Como el estadístico de prueba Z ({hypothetical_z_reject:.4f}) es >= valor crítico superior ({critical_value_upper:.4f}), rechazamos la hipótesis nula.")
print("""**Conclusión Práctica:** Hay suficiente evidencia, al nivel de significancia de 5%, para concluir que la media de la población es significativamente diferente de 50.

Recuerde: Una decisión de 'no rechazar H0' no prueba que H0 sea verdadera. Simplemente significa que no encontramos evidencia suficiente en la muestra para demostrar que es falsa al nivel de significancia elegido.
""")


# @title 4. Intervalos de Confianza: Cálculo y Interpretación
# @markdown Comprende qué son los intervalos de confianza y cómo calcularlos e interpretarlos.

print("""## Intervalos de Confianza: Cálculo y Interpretación

Un intervalo de confianza es un rango de valores que probablemente contenga un parámetro de población desconocido con un cierto nivel de confianza. Cuantifica la incertidumbre de una estimación puntual.

### El Nivel de Confianza:

*   La probabilidad de que, si se repitiera el muestreo y la construcción del intervalo muchas veces, el intervalo resultante contenga el verdadero parámetro de la población.
*   No es la probabilidad de que el parámetro poblacional esté en un intervalo específico calculado.

### Cálculo del Intervalo de Confianza para la Media Poblacional (con σ conocida)

Usando los mismos datos simulados: Media muestral (x̄) = 53, Desviación estándar poblacional (σ) = 10, Tamaño de muestra (n) = 30. Calcularemos un intervalo de confianza del 95% para la media poblacional (μ).

Fórmula: x̄ ± Z * (σ / √n)
""")

# @title Calcular y Mostrar Intervalo de Confianza
# @markdown Calcula el intervalo de confianza usando la fórmula manual y la función de SciPy.

# Usando los datos simulados: x_bar, sigma, n del paso anterior
# Nivel de confianza del 95%
confidence_level = 0.95
alpha_ci = 1 - confidence_level # Alpha para el intervalo de confianza
print(f"Calculando el intervalo de confianza para la media poblacional (μ) con:")
print(f"  Media muestral (x̄): {x_bar}")
print(f"  Desviación estándar poblacional (σ): {sigma}")
print(f"  Tamaño de muestra (n): {n}")
print(f"  Nivel de confianza: {confidence_level*100}%\n")


# Para desviación estándar poblacional conocida (sigma) y tamaño de muestra >= 30,
# usamos la distribución Z.
# Encontrar el valor Z crítico para el nivel de confianza del 95% (bilateral)
# Usamos norm.ppf(1 - alpha_ci / 2) para encontrar el Z que deja alpha_ci/2 en la cola superior.
z_critical_ci = norm.ppf(1 - alpha_ci / 2)

# Calcular el margen de error
margin_of_error = z_critical_ci * (sigma / np.sqrt(n))

# Calcular los límites del intervalo de confianza
lower_bound_ci = x_bar - margin_of_error
upper_bound_ci = x_bar + margin_of_error

# Imprimir el Intervalo de Confianza Calculado
print(f"El valor crítico Z para un nivel de confianza del {confidence_level*100}% es aproximadamente {z_critical_ci:.4f}")
print(f"El margen de error es: {margin_of_error:.4f}")
print(f"El intervalo de confianza del {confidence_level*100}% para la media poblacional es: [{lower_bound_ci:.4f}, {upper_bound_ci:.4f}]\n")


# Usar scipy.stats.norm.interval como alternativa (verificación)
# Parámetros: nivel de confianza, loc (media), scale (error estándar = sigma / sqrt(n))
confidence_interval_scipy = norm.interval(confidence_level, loc=x_bar, scale=sigma / np.sqrt(n))

print(f"Intervalo de confianza calculado usando scipy.stats.norm.interval: {confidence_interval_scipy}\n")


print("""### Explicación de los Parámetros Utilizados:

*   **Media muestral (x̄):** La media de los datos observados en nuestra muestra. Es la mejor estimación puntual del parámetro de la población.
*   **Desviación estándar poblacional (σ):** La desviación estándar de la población completa (supuesta conocida). En realidad, si no se conoce, se usa la desviación estándar muestral (s) y la distribución t de Student.
*   **Tamaño de muestra (n):** El número de observaciones en nuestra muestra. Un 'n' mayor reduce el error estándar (σ/√n), resultando en un intervalo más estrecho y preciso.
*   **Nivel de confianza:** El grado de seguridad deseado. Un nivel más alto requiere un valor crítico Z mayor y un intervalo más amplio.
*   **Valor crítico Z:** Número de errores estándar desde la media muestral. Depende del nivel de confianza y la distribución.

### Interpretación del Intervalo de Confianza:
""")
print(f"""Basado en nuestra muestra, tenemos un {confidence_level*100}% de confianza en que el verdadero valor de la media poblacional (μ) se encuentra entre {lower_bound_ci:.4f} y {upper_bound_ci:.4f}.

Esto significa que si repitiéramos este estudio muchas veces, el 95% de los intervalos de confianza construidos de esta manera contendrían la verdadera media poblacional. La probabilidad se refiere al método, no a un intervalo específico. El intervalo nos da un rango de valores plausibles para μ.
""")


# @title 5. Cálculo del Tamaño de Muestra
# @markdown Aprende por qué es importante el tamaño de muestra y cómo calcularlo para diferentes propósitos estadísticos.

print("""## Cálculo del Tamaño de Muestra

Determinar el tamaño de muestra apropiado es crucial para la validez y eficiencia de un estudio.

### Importancia del Tamaño de Muestra:

*   **Precisión:** Un 'n' más grande mejora la precisión de las estimaciones.
*   **Poder:** Un 'n' adecuado aumenta la probabilidad de detectar efectos reales en pruebas de hipótesis (poder estadístico).
*   **Generalización:** Permite generalizar los hallazgos de la muestra a la población.
*   **Eficiencia:** Optimiza el uso de recursos.

### Factores que Influyen en el Tamaño de Muestra:

*   **Margen de Error Deseado (E):** Tolerancia de diferencia entre la estimación muestral y el parámetro poblacional. Menor E requiere mayor 'n'.
*   **Nivel de Confianza Deseado:** Mayor confianza (ej. 99% vs 95%) requiere mayor 'n'.
*   **Variabilidad Poblacional (σ o s):** Mayor variabilidad requiere mayor 'n'.
*   **Poder Deseado (1 - β):** Mayor poder (ej. 90% vs 80%) requiere mayor 'n' (para pruebas de hipótesis).
*   **Nivel de Significancia (α):** Alpha más pequeño (ej. 0.01 vs 0.05) puede requerir mayor 'n' (para mantener el poder).
*   **Tamaño del Efecto (d):** Magnitud de la diferencia a detectar. Un efecto más pequeño requiere mayor 'n'.

### Cálculo para Estimar una Media Poblacional

Objetivo: Estimar la media poblacional con un cierto nivel de confianza y un margen de error máximo aceptable.

Fórmula: n = (Z * σ / E)²
""")

# @title Calcular Tamaño de Muestra para Estimación de Media
# @markdown Calcula el tamaño de muestra requerido para estimar la media poblacional bajo diferentes escenarios.

# Fórmula para tamaño de muestra para estimar media: n = (Z * σ / E)^2
print("Fórmula utilizada: n = (Z * σ / E)^2")
print("Donde:")
print("  n = Tamaño de muestra requerido")
print("  Z = Puntuación Z para el nivel de confianza deseado")
print("  σ = Desviación estándar de la población (o una estimación)")
print("  E = Margen de error máximo aceptable\n")


# Ejemplo de Cálculo 1: Estimar media con 95% confianza, margen de error 5, desv. estándar 10
desired_confidence_level = 0.95
desired_margin_of_error = 5
estimated_sigma = 10 # Necesitamos una estimación de la desviación estándar poblacional

# Encontrar la puntuación Z para el nivel de confianza deseado (bilateral)
alpha_ss = 1 - desired_confidence_level
z_score_estimate = norm.ppf(1 - alpha_ss / 2) # Z para el percentil (1 - alpha/2)

# Calcular el tamaño de muestra requerido
# Usamos math.ceil para redondear hacia arriba, ya que el tamaño de muestra debe ser un entero.
n_estimate_mean = math.ceil((z_score_estimate * estimated_sigma / desired_margin_of_error) ** 2)

print(f"Ejemplo 1: Estimar la media poblacional")
print(f"  Nivel de confianza deseado: {desired_confidence_level*100}%")
print(f"  Margen de error deseado (E): {desired_margin_of_error}")
print(f"  Desviación estándar estimada (σ): {estimated_sigma}")
print(f"  Puntuación Z para {desired_confidence_level*100}%: {z_score_estimate:.4f}")
print(f"  Tamaño de muestra requerido: ({z_score_estimate:.4f} * {estimated_sigma} / {desired_margin_of_error})^2 = {n_estimate_mean}\n")


# Ejemplo de Cálculo 2: Parámetros diferentes
desired_confidence_level_ex2 = 0.99
desired_margin_of_error_ex2 = 2
estimated_sigma_ex2 = 15

alpha_ss_ex2 = 1 - desired_confidence_level_ex2
z_score_estimate_ex2 = norm.ppf(1 - alpha_ss_ex2 / 2)
n_estimate_mean_ex2 = math.ceil((z_score_estimate_ex2 * estimated_sigma_ex2 / desired_margin_of_error_ex2) ** 2)

print(f"Ejemplo 2: Estimar la media poblacional con parámetros diferentes")
print(f"  Nivel de confianza deseado: {desired_confidence_level_ex2*100}%")
print(f"  Margen de error deseado (E): {desired_margin_of_error_ex2}")
print(f"  Desviación estándar estimada (σ): {estimated_sigma_ex2}")
print(f"  Puntuación Z para {desired_confidence_level_ex2*100}%: {z_score_estimate_ex2:.4f}")
print(f"  Tamaño de muestra requerido: ({z_score_estimate_ex2:.4f} * {estimated_sigma_ex2} / {desired_margin_of_error_ex2})^2 = {n_estimate_mean_ex2}\n")

print("""**Observaciones:**
- Un margen de error más pequeño (Ejemplo 2 vs Ejemplo 1) requiere un tamaño de muestra significativamente mayor.
- Un nivel de confianza más alto (Ejemplo 2 vs Ejemplo 1) requiere un tamaño de muestra mayor.
- Una mayor variabilidad (σ) en la población también requiere un tamaño de muestra mayor.
""")

# @title Calcular Tamaño de Muestra para Prueba de Hipótesis (Test Z Unilateral)
# @markdown Calcula el tamaño de muestra requerido para realizar una prueba de hipótesis con un poder y nivel de significancia deseados.

print("""### Cálculo para una Prueba de Hipótesis (Test Z Unilateral para la Media)

Objetivo: Determinar el tamaño de muestra necesario para detectar una diferencia mínima (tamaño del efecto) con un poder y nivel de significancia especificados.

Fórmula utilizada (Test Unilateral, σ conocida): n = ((Z_α + Z_β) * σ / δ)²
""")

# Fórmula para tamaño de muestra (test unilateral, sigma conocida):
# n = ((Z_alpha + Z_beta) * σ / δ)^2
print("Fórmula utilizada (Test Unilateral): n = ((Z_α + Z_β) * σ / δ)^2")
print("Donde:")
print("  n = Tamaño de muestra requerido")
print("  Z_α = Puntuación Z para el nivel de significancia (α)")
print("  Z_β = Puntuación Z para el poder deseado (1 - β). Beta (β) es la probabilidad de un error Tipo II.")
print("  σ = Desviación estándar de la población (o una estimación previa)")
print("  δ = Diferencia mínima que se desea detectar (Tamaño del efecto en unidades originales)\n")

# Nota: Para un test bilateral, la fórmula usa Z_(alpha/2).
# La fórmula también se puede expresar en términos de tamaño del efecto estandarizado (Cohen's d = δ / σ):
# n = ((Z_alpha + Z_beta) / d)^2
print("Alternativamente, usando el tamaño del efecto estandarizado (d = δ / σ): n = ((Z_α + Z_β) / d)^2\n")


# Ejemplo de Cálculo 1: Test Z Unilateral
# Queremos detectar una diferencia de 3 unidades de la media nula (e.g., H0: mu=50, queremos detectar si mu=53 o más)
# Nivel de significancia (alpha): 0.05 (test unilateral)
# Poder deseado: 0.80 (beta = 0.20)
# Desviación estándar poblacional estimada (sigma): 10
# Diferencia mínima a detectar (delta): 3

significance_level = 0.05
desired_power = 0.80
estimated_sigma = 10
minimum_detectable_difference = 3 # delta

# Encontrar la puntuación Z para alpha (para un test unilateral, usamos 1-alpha)
z_alpha = norm.ppf(1 - significance_level)

# Encontrar la puntuación Z para beta (para calcular el poder, usamos 1-beta)
beta = 1 - desired_power
z_beta = norm.ppf(1 - beta) # Este es el Z-score correspondiente al percentil del poder deseado

# Calcular el tamaño de muestra requerido
n_hypothesis_test = math.ceil(((z_alpha + z_beta) * estimated_sigma / minimum_detectable_difference) ** 2)

print(f"Ejemplo 1: Prueba de Hipótesis (Test Z Unilateral)")
print(f"  Nivel de significancia (α): {significance_level}")
print(f"  Poder deseado (1 - β): {desired_power}")
print(f"  Desviación estándar estimada (σ): {estimated_sigma}")
print(f"  Diferencia mínima a detectar (δ): {minimum_detectable_difference}")
print(f"  Puntuación Z para α ({significance_level}): {z_alpha:.4f}")
print(f"  Puntuación Z para 1-β ({desired_power}): {z_beta:.4f}")
print(f"  Tamaño de muestra requerido: (({z_alpha:.4f} + {z_beta:.4f}) * {estimated_sigma} / {minimum_detectable_difference})^2 = {n_hypothesis_test}\n")


# Otro Ejemplo: Parámetros diferentes
# Nivel de significancia (alpha): 0.01 (unilateral)
# Poder deseado: 0.95
# Desviación estándar poblacional estimada (sigma): 15
# Diferencia mínima a detectar (delta): 2

significance_level_ex2 = 0.01
desired_power_ex2 = 0.95
estimated_sigma_ex2 = 15
minimum_detectable_difference_ex2 = 2

z_alpha_ex2 = norm.ppf(1 - significance_level_ex2)
beta_ex2 = 1 - desired_power_ex2
z_beta_ex2 = norm.ppf(1 - beta_ex2)
n_hypothesis_test_ex2 = math.ceil(((z_alpha_ex2 + z_beta_ex2) * estimated_sigma_ex2 / minimum_detectable_difference_ex2) ** 2)


print(f"Otro Ejemplo: Prueba de Hipótesis (Test Z Unilateral) con parámetros diferentes")
print(f"  Nivel de significancia (α): {significance_level_ex2}")
print(f"  Poder deseado (1 - β): {desired_power_ex2}")
print(f"  Desviación estándar estimada (σ): {estimated_sigma_ex2}")
print(f"  Diferencia mínima a detectar (δ): {minimum_detectable_difference_ex2}")
print(f"  Puntuación Z para α ({significance_level_ex2}): {z_alpha_ex2:.4f}")
print(f"  Puntuación Z para 1-β ({desired_power_ex2}): {z_beta_ex2:.4f}")
print(f"  Tamaño de muestra requerido: (({z_alpha_ex2:.4f} + {z_beta_ex2:.4f}) * {estimated_sigma_ex2} / {minimum_detectable_difference_ex2})^2 = {n_hypothesis_test_ex2}\n")

print("""**Observaciones:**
- Un poder deseado más alto (Ejemplo 2 vs Ejemplo 1) requiere un tamaño de muestra mayor.
- Un nivel de significancia más estricto (alpha más pequeño, Ejemplo 2 vs Ejemplo 1) requiere un tamaño de muestra mayor.
- La detección de una diferencia más pequeña (delta más pequeño, Ejemplo 2 vs Ejemplo 1) requiere un tamaño de muestra significativamente mayor.
- Una mayor variabilidad (σ) también requiere un tamaño de muestra mayor.
""")

# @title Cálculo del Tamaño de Muestra usando statsmodels (Opcional)
# @markdown statsmodels es una librería más completa para análisis de poder y cálculo de tamaño de muestra.

if _STATMODELS_AVAILABLE:
    print("""### Cálculo del Tamaño de Muestra usando statsmodels (Alternativa)

La librería `statsmodels` ofrece funciones más avanzadas para el análisis de poder y el cálculo del tamaño de muestra para una variedad de pruebas.

Usaremos `statsmodels.stats.power.NormalIndPower` para calcular el tamaño de muestra para un test Z unilateral para la media. Requiere especificar el **tamaño del efecto estandarizado (Cohen's d = δ / σ)**.
""")

    # Ejemplo usando statsmodels para test Z unilateral para la media
    # Para el primer ejemplo: delta=3, sigma=10, entonces Cohen's d = 3/10 = 0.3 (tamaño del efecto medio)

    effect_size_statmodels = minimum_detectable_difference / estimated_sigma
    alpha_statmodels = significance_level
    power_statmodels = desired_power
    ratio = 0 # ratio de tamaños de muestra, 0 para test de una muestra

    # Crear un objeto de análisis de poder para distribución normal (test Z)
    power_analysis = NormalIndPower()

    # Calcular tamaño de muestra (para un test de una muestra, es el total)
    # 'solve_for' = 'nobs1' calcula el tamaño de muestra para el primer grupo (nuestra única muestra)
    # 'alternative' especifica el tipo de test ('two-sided', 'larger' o 'smaller')
    n_statmodels = power_analysis.solve_power(
        effect_size=effect_size_statmodels,
        alpha=alpha_statmodels,
        power=power_statmodels,
        ratio=ratio, # Usar ratio=0 para test de una muestra
        alternative='larger' # Opciones válidas: 'two-sided', 'larger', 'smaller'. Usamos 'larger' asumiendo el ejemplo unilateral.
    )

    print(f"Ejemplo usando statsmodels (Test Z Unilateral para la Media, alternativa='larger'):")
    print(f"  Tamaño del efecto (Cohen's d = δ/σ): {effect_size_statmodels:.4f}")
    print(f"  Nivel de significancia (α): {alpha_statmodels}")
    print(f"  Poder deseado (1 - β): {power_statmodels}")
    print(f"  Tamaño de muestra requerido: {math.ceil(n_statmodels)}\n")


    # Ejemplo usando statsmodels para el segundo escenario
    effect_size_statmodels_ex2 = minimum_detectable_difference_ex2 / estimated_sigma_ex2
    alpha_statmodels_ex2 = significance_level_ex2
    power_statmodels_ex2 = desired_power_ex2

    n_statmodels_ex2 = power_analysis.solve_power(
        effect_size=effect_size_statmodels_ex2,
        alpha=alpha_statmodels_ex2,
        power=power_statmodels_ex2,
        ratio=ratio,
        alternative='larger' # Usamos 'larger' asumiendo el ejemplo unilateral.
    )
    print(f"Otro Ejemplo usando statsmodels (Test Z Unilateral, alternativa='larger') con parámetros diferentes:")
    print(f"  Tamaño del efecto (Cohen's d = δ/σ): {effect_size_statmodels_ex2:.4f}")
    print(f"  Nivel de significancia (α): {alpha_statmodels_ex2}")
    print(f"  Poder deseado (1 - β): {power_statmodels_ex2}")
    print(f"  Tamaño de muestra requerido: {math.ceil(n_statmodels_ex2)}\n")

else:
    print("La librería statsmodels no está disponible. Por favor, instálala para ejecutar esta sección.")


print("""### Explicación de los Parámetros para Tamaño de Muestra (Pruebas de Hipótesis):

*   **Nivel de Significancia (α):** Probabilidad de Error Tipo I. Un α más pequeño requiere mayor 'n' para mantener el poder.
*   **Poder (1 - β):** Probabilidad de Error Tipo II (β). Mayor poder requiere mayor 'n'.
*   **Desviación Estándar (σ):** Variabilidad. Mayor σ requiere mayor 'n'.
*   **Diferencia Mínima a Detectar (δ) / Tamaño del Efecto (d):** Magnitud del efecto a detectar. Un efecto más pequeño requiere mayor 'n'.
*   **Z_α y Z_β:** Valores Z de la distribución normal estándar asociados a alpha y beta.

**Consideraciones adicionales:**
- La elección de test unilateral vs bilateral afecta el valor Z_α.
- Si σ es desconocida, se usa una estimación previa.
- `statsmodels` es útil para casos más complejos y diferentes tipos de pruebas.
""")
```

```
# -*- coding: utf-8 -*-
"""Cuaderno de Google Colab: Pruebas de Hipótesis, Intervalos de Confianza y Tamaño de Muestra

Este cuaderno interactivo explora y demuestra los conceptos fundamentales
de las pruebas de hipótesis, los intervalos de confianza y la determinación
del tamaño de muestra en estadística, utilizando Python y librerías comunes
como NumPy, SciPy y Matplotlib.

Contenido:
1.  Introducción a los conceptos.
2.  Pruebas de Hipótesis: pasos, cálculo e interpretación.
3.  Intervalos de Confianza: cálculo e interpretación.
4.  Tamaño de Muestra: cálculo para estimación y pruebas de hipótesis.
"""

# @title 1. Instalación e Importación de Librerías
# @markdown Ejecuta esta celda para instalar e importar las librerías necesarias.

# Instalación de librerías (si no están preinstaladas en Colab)
# !pip install numpy scipy matplotlib statsmodels

# Importación de librerías
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm # Usado para distribuciones Normales (Z)
import math # Usado para funciones matemáticas como ceil
try:
    from statsmodels.stats.power import NormalIndPower # Usado para análisis de poder y tamaño de muestra con statsmodels
    _STATMODELS_AVAILABLE = True
except ImportError:
    _STATMODELS_AVAILABLE = False
    print("Advertencia: La librería statsmodels no está instalada. Algunas secciones de cálculo de tamaño de muestra no funcionarán.")


# Configuración para matplotlib (inline en cuadernos)
# %matplotlib inline

print("Librerías instaladas e importadas.")

# @title 2. Introducción a los Conceptos Estadísticos
# @markdown Lee esta sección para entender la base de las pruebas de hipótesis, intervalos de confianza y tamaño de muestra.

print("""# Introducción a las Pruebas de Hipótesis, Intervalos de Confianza y Tamaño de Muestra

En el análisis de datos y la inferencia estadística, utilizamos herramientas para tomar decisiones informadas sobre una población basándonos en una muestra.

*   **Pruebas de Hipótesis:** Un método formal para determinar si hay suficiente evidencia en una muestra de datos para rechazar una hipótesis nula sobre una población. Nos ayudan a evaluar si una observación es estadísticamente significativa o si podría deberse simplemente al azar.

*   **Intervalos de Confianza:** Proporcionan un rango de valores dentro del cual es probable que se encuentre un parámetro de población desconocido. Nos dan una idea de la precisión de nuestra estimación basada en la muestra.

*   **Tamaño de la Muestra:** Se refiere al número de observaciones o individuos incluidos en un estudio. Determinar un tamaño de muestra adecuado es crucial porque un tamaño de muestra insuficiente puede llevar a conclusiones poco fiables, mientras que uno excesivamente grande puede ser ineficiente en términos de recursos.

Estos conceptos son fundamentales en la práctica porque nos permiten sacar conclusiones fiables de los datos, cuantificar la incertidumbre de nuestras estimaciones y diseñar estudios eficientes que maximicen la probabilidad de detectar efectos reales si existen.
""")


# @title 3. Pruebas de Hipótesis: Pasos y Demostración
# @markdown Explora los 7 pasos de una prueba de hipótesis y visualiza los conceptos clave como la región de rechazo y el valor p.

print("""## Pruebas de Hipótesis: Pasos y Demostración

Una prueba de hipótesis sigue una serie de pasos para evaluar una afirmación sobre un parámetro poblacional. Aquí detallamos los pasos y usamos un ejemplo para ilustrarlos.

### Pasos de una Prueba de Hipótesis:

**Paso 1: Definir las Hipótesis**
*   **Hipótesis Nula (H0):** La afirmación que se asume como verdadera hasta que haya evidencia suficiente en su contra (ej. no hay efecto, no hay diferencia).
*   **Hipótesis Alternativa (H1):** La afirmación que se desea probar; lo que se concluye si se rechaza H0 (ej. hay un efecto, hay una diferencia).

**Paso 2: Establecer el Nivel de Significancia (Alpha - α)**
*   La probabilidad máxima permitida de cometer un Error de Tipo I (rechazar H0 cuando es verdadera). Comúnmente 0.05.

**Paso 3: Identificar el Estadístico de Prueba**
*   Un valor calculado a partir de la muestra que mide cuánto se desvían los datos de lo esperado bajo H0 (ej. estadístico Z, t, F).

**Paso 4: Determinar el Valor Crítico y la Región de Rechazo**
*   **Valor Crítico:** El umbral en la distribución del estadístico de prueba que separa la región de no rechazo de la región de rechazo.
*   **Región de Rechazo:** El conjunto de valores del estadístico de prueba que llevan a rechazar H0.

**Paso 5: Calcular el Estadístico de Prueba y el Valor p**
*   Calcular el estadístico de prueba usando los datos de la muestra.
*   **Valor p:** La probabilidad de obtener un estadístico de prueba tan extremo o más extremo que el observado, asumiendo que H0 es verdadera.

**Paso 6: Tomar una Decisión Estadística**
*   Comparar el estadístico de prueba con el valor crítico, o el valor p con alpha. Si el estadístico de prueba cae en la región de rechazo (o p-value <= alpha), se rechaza H0.

**Paso 7: Extraer una Conclusión Práctica**
*   Interpretar la decisión estadística en el contexto del problema. Indicar si hay suficiente evidencia para apoyar H1 o no hay suficiente evidencia para rechazar H0.

### Demostración con Código (Test Z Bilateral para la Media)

Consideremos un ejemplo hipotético: queremos probar si la media de una población es diferente de 50, con un nivel de significancia del 5%. Asumimos que conocemos la desviación estándar poblacional (σ = 10).
""")

# @title Paso 1-4: Hipótesis, Alpha, Valor Crítico y Región de Rechazo
# @markdown Define las hipótesis, el nivel de significancia y calcula/visualiza los valores críticos y la región de rechazo para un test Z bilateral.

# Paso 1: Definir las Hipótesis
# H0: μ = 50
# H1: μ ≠ 50 (Test Bilateral)

# Paso 2: Establecer el Nivel de Significancia (Alpha)
alpha = 0.05
print(f"Nivel de significancia (alpha): {alpha}")

# Paso 4: Determinar Valores Críticos y Región de Rechazo para un test bilateral Z
# Para un test Z bilateral, alpha se divide en las dos colas de la distribución.
# Usamos norm.ppf (inverse CDF) para encontrar el valor Z que deja alpha/2 en cada cola.
critical_value_lower = norm.ppf(alpha / 2)
critical_value_upper = norm.ppf(1 - alpha / 2)

print(f"Valor crítico inferior (Z) para alpha/2: {critical_value_lower:.4f}")
print(f"Valor crítico superior (Z) para 1-alpha/2: {critical_value_upper:.4f}")
print(f"Región de rechazo: Z < {critical_value_lower:.4f} o Z > {critical_value_upper:.4f}\n")


# Graficar la distribución normal estándar y sombrear la región de rechazo (Visualización Paso 4)
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1) # Distribución Normal Estándar (media=0, desv. estándar=1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Normal Estándar (Z)')

# Sombrear la región de rechazo inferior
x_reject_lower = x[x < critical_value_lower]
y_reject_lower = norm.pdf(x_reject_lower, 0, 1)
plt.fill_between(x_reject_lower, 0, y_reject_lower, color='red', alpha=0.5, label='Región de Rechazo (α/2)')

# Sombrear la región de rechazo superior
x_reject_upper = x[x > critical_value_upper]
y_reject_upper = norm.pdf(x_reject_upper, 0, 1)
plt.fill_between(x_reject_upper, 0, y_reject_upper, color='red', alpha=0.5) # No se necesita etiqueta duplicada


plt.title('Distribución Normal Estándar con Región de Rechazo (Test Bilateral)')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.axvline(critical_value_lower, color='black', linestyle='--', label=f'Valor Crítico Inferior ({critical_value_lower:.2f})')
plt.axvline(critical_value_upper, color='black', linestyle='--', label=f'Valor Crítico Superior ({critical_value_upper:.2f})')
plt.legend()
plt.grid(True)
plt.show()

print("""La gráfica muestra la distribución normal estándar. Las áreas sombreadas en rojo representan la región de rechazo para un test bilateral con alpha = 0.05.
Si el estadístico de prueba Z calculado a partir de los datos de la muestra cae dentro de estas áreas sombreadas,
rechazaríamos la hipótesis nula (H0) al nivel de significancia de 5%.
""")

# @title Paso 5-6: Calcular Estadístico de Prueba, Valor p y Graficar Resultados
# @markdown Simula datos de muestra, calcula el estadístico de prueba y el valor p, y grafica su posición relative a la región de rechazo.

# Simular datos de muestra para nuestro escenario hipotético
n = 30      # Tamaño de la muestra
x_bar = 53  # Media de la muestra observada
mu_0 = 50   # Media poblacional bajo H0 (de H0: μ = 50)
sigma = 10  # Desviación estándar poblacional (supuesta conocida)

print(f"Escenario simulado: Tamaño de muestra n = {n}, Media de la muestra x̄ = {x_bar}")
print(f"Hipótesis Nula H0: μ = {mu_0}, Desviación estándar de la población σ = {sigma}\n")

# Paso 5: Calcular el Estadístico de Prueba Z
# Fórmula para Test Z de una muestra con sigma conocida: Z = (x_bar - mu_0) / (sigma / sqrt(n))
z_statistic = (x_bar - mu_0) / (sigma / np.sqrt(n))

# Calcular el valor p (para un test bilateral)
# El valor p es la probabilidad de observar un estadístico Z tan extremo como el calculado, en cualquier cola.
# Usamos norm.sf (función de supervivencia = 1 - CDF) para la cola superior y multiplicamos por 2 para bilateral.
p_value = 2 * norm.sf(abs(z_statistic))

# Imprimir el estadístico de prueba calculado y el valor p
print(f"Estadístico de Prueba Z calculado: {z_statistic:.4f}")
print(f"Valor p calculado (bilateral): {p_value:.4f}\n")


# Paso 6: Graficar los resultados (distribución, valores críticos, estadístico de prueba)
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1) # Distribución Normal Estándar

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Normal Estándar (Z)')

# Sombrear la región de rechazo (usando los valores críticos calculados previamente)
x_reject_lower = x[x < critical_value_lower]
y_reject_lower = norm.pdf(x_reject_lower, 0, 1)
plt.fill_between(x_reject_lower, 0, y_reject_lower, color='red', alpha=0.5, label='Región de Rechazo (α/2)')

x_reject_upper = x[x > critical_value_upper]
y_reject_upper = norm.pdf(x_reject_upper, 0, 1)
plt.fill_between(x_reject_upper, 0, y_reject_upper, color='red', alpha=0.5)


# Agregar líneas verticales para los valores críticos y el estadístico de prueba
plt.axvline(critical_value_lower, color='black', linestyle='--', label=f'Valor Crítico Inferior ({critical_value_lower:.2f})')
plt.axvline(critical_value_upper, color='black', linestyle='--', label=f'Valor Crítico Superior ({critical_value_upper:.2f})')
plt.axvline(z_statistic, color='green', linestyle='-', label=f'Estadístico de Prueba Z ({z_statistic:.2f})', linewidth=2)


plt.title('Distribución Normal Estándar con Región de Rechazo, Valores Críticos y Estadístico de Prueba')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

print("""La gráfica visualiza el estadístico de prueba Z calculado (línea verde) en relación con los valores críticos (líneas negras) y la región de rechazo (áreas rojas).
Si la línea verde cae en las áreas rojas, rechazamos la hipótesis nula.
""")

# @title Paso 7: Decisión Estadística y Conclusión Práctica
# @markdown Interpreta los resultados de la prueba de hipótesis basándose en el estadístico de prueba, el valor p y el nivel de significancia.

print("""## Paso 7: Decisión Estadística y Conclusión Práctica

Una vez calculados el estadístico de prueba y el valor p, tomamos una decisión estadística y la traducimos en una conclusión práctica.

### Decisión Estadística:

*   **Usando el Valor Crítico:** Rechazamos H0 si el estadístico de prueba cae en la región de rechazo (Z <= critical_value_lower o Z >= critical_value_upper para un test bilateral).
*   **Usando el Valor p:** Rechazamos H0 si el valor p es menor o igual que alpha (p-value <= alpha).

### Conclusión Práctica:

*   Traduce la decisión estadística al contexto del problema. Evita decir "aceptar H0"; si no se rechaza H0, simplemente significa que no hay suficiente evidencia en la muestra para rechazarla al nivel de significancia elegido.

### Significado e Interpretación del Valor p:

*   El valor p es la probabilidad de observar los datos (o resultados más extremos) si H0 fuera verdadera.
*   Un valor p pequeño (< alpha) sugiere que los datos son improbables bajo H0, proporcionando evidencia contra H0.
*   Un valor p grande (> alpha) sugiere que los datos son razonablemente probables bajo H0, por lo que no hay evidencia suficiente para rechazar H0.

### Ejemplos de Interpretación (basado en nuestro ejemplo):
""")

print(f"Nuestro estadístico de prueba Z calculado fue {z_statistic:.4f}.")
print(f"Nuestro valor p calculado fue {p_value:.4f}.")
print(f"Nuestro nivel de significancia alpha fue {alpha}.\n")

# Escenario 1: Basado en nuestros resultados actuales (no rechazar H0)
print("#### Escenario de nuestro Ejemplo:")
print(f"**Decisión Estadística (usando el valor p):** Como el valor p ({p_value:.4f}) es > alpha ({alpha}), no rechazamos la hipótesis nula.")
print(f"**Decisión Estadística (usando el valor crítico):** Como el estadístico de prueba Z ({z_statistic:.4f}) está entre los valores críticos ({critical_value_lower:.4f} y {critical_value_upper:.4f}), no rechazamos la hipótesis nula.")
print("""**Conclusión Práctica:** No hay suficiente evidencia, al nivel de significancia de 5%, para concluir que la media de la población es significativamente diferente de 50. Esto no significa que la media sea exactamente 50, solo que la evidencia de esta muestra no es lo suficientemente fuerte para afirmar que es diferente.

""")

# Escenario 2: Caso hipotético donde H0 es rechazado
hypothetical_z_reject = 2.5 # Supongamos un Estadístico Z de 2.5
hypothetical_p_reject = 2 * norm.sf(abs(hypothetical_z_reject)) # Calcular el p-value para este Z hipotético

print("#### Escenario Hipotético (donde se rechazaría H0):")
print(f"Supongamos que el estadístico de prueba Z fuera {hypothetical_z_reject:.4f} y el valor p fuera {hypothetical_p_reject:.4f}.")
print(f"**Decisión Estadística (usando el valor p):** Como el valor p ({hypothetical_p_reject:.4f}) es <= alpha ({alpha}), rechazamos la hipótesis nula.")
print(f"**Decisión Estadística (usando el valor crítico):** Como el estadístico de prueba Z ({hypothetical_z_reject:.4f}) es >= valor crítico superior ({critical_value_upper:.4f}), rechazamos la hipótesis nula.")
print("""**Conclusión Práctica:** Hay suficiente evidencia, al nivel de significancia de 5%, para concluir que la media de la población es significativamente diferente de 50.

Recuerde: Una decisión de 'no rechazar H0' no prueba que H0 sea verdadera. Simplemente significa que no encontramos evidencia suficiente en la muestra para demostrar que es falsa al nivel de significancia elegido.
""")


# @title 4. Intervalos de Confianza: Cálculo y Interpretación
# @markdown Comprende qué son los intervalos de confianza y cómo calcularlos e interpretarlos.

print("""## Intervalos de Confianza: Cálculo y Interpretación

Un intervalo de confianza es un rango de valores que probablemente contenga un parámetro de población desconocido con un cierto nivel de confianza. Cuantifica la incertidumbre de una estimación puntual.

### El Nivel de Confianza:

*   La probabilidad de que, si se repitiera el muestreo y la construcción del intervalo muchas veces, el intervalo resultante contenga el verdadero parámetro de la población.
*   No es la probabilidad de que el parámetro poblacional esté en un intervalo específico calculado.

### Cálculo del Intervalo de Confianza para la Media Poblacional (con σ conocida)

Usando los mismos datos simulados: Media muestral (x̄) = 53, Desviación estándar poblacional (σ) = 10, Tamaño de muestra (n) = 30. Calcularemos un intervalo de confianza del 95% para la media poblacional (μ).

Fórmula: x̄ ± Z * (σ / √n)
""")

# @title Calcular y Mostrar Intervalo de Confianza
# @markdown Calcula el intervalo de confianza usando la fórmula manual y la función de SciPy.

# Usando los datos simulados: x_bar, sigma, n del paso anterior
# Nivel de confianza del 95%
confidence_level = 0.95
alpha_ci = 1 - confidence_level # Alpha para el intervalo de confianza
print(f"Calculando el intervalo de confianza para la media poblacional (μ) con:")
print(f"  Media muestral (x̄): {x_bar}")
print(f"  Desviación estándar poblacional (σ): {sigma}")
print(f"  Tamaño de muestra (n): {n}")
print(f"  Nivel de confianza: {confidence_level*100}%\n")


# Para desviación estándar poblacional conocida (sigma) y tamaño de muestra >= 30,
# usamos la distribución Z.
# Encontrar el valor Z crítico para el nivel de confianza del 95% (bilateral)
# Usamos norm.ppf(1 - alpha_ci / 2) para encontrar el Z que deja alpha_ci/2 en la cola superior.
z_critical_ci = norm.ppf(1 - alpha_ci / 2)

# Calcular el margen de error
margin_of_error = z_critical_ci * (sigma / np.sqrt(n))

# Calcular los límites del intervalo de confianza
lower_bound_ci = x_bar - margin_of_error
upper_bound_ci = x_bar + margin_of_error

# Imprimir el Intervalo de Confianza Calculado
print(f"El valor crítico Z para un nivel de confianza del {confidence_level*100}% es aproximadamente {z_critical_ci:.4f}")
print(f"El margen de error es: {margin_of_error:.4f}")
print(f"El intervalo de confianza del {confidence_level*100}% para la media poblacional es: [{lower_bound_ci:.4f}, {upper_bound_ci:.4f}]\n")


# Usar scipy.stats.norm.interval como alternativa (verificación)
# Parámetros: nivel de confianza, loc (media), scale (error estándar = sigma / sqrt(n))
confidence_interval_scipy = norm.interval(confidence_level, loc=x_bar, scale=sigma / np.sqrt(n))

print(f"Intervalo de confianza calculado usando scipy.stats.norm.interval: {confidence_interval_scipy}\n")


print("""### Explicación de los Parámetros Utilizados:

*   **Media muestral (x̄):** La media de los datos observados en nuestra muestra. Es la mejor estimación puntual del parámetro de la población.
*   **Desviación estándar poblacional (σ):** La desviación estándar de la población completa (supuesta conocida). En realidad, si no se conoce, se usa la desviación estándar muestral (s) y la distribución t de Student.
*   **Tamaño de muestra (n):** El número de observaciones en nuestra muestra. Un 'n' mayor reduce el error estándar (σ/√n), resultando en un intervalo más estrecho y preciso.
*   **Nivel de confianza:** El grado de seguridad deseado. Un nivel más alto requiere un valor crítico Z mayor y un intervalo más amplio.
*   **Valor crítico Z:** Número de errores estándar desde la media muestral. Depende del nivel de confianza y la distribución.

### Interpretación del Intervalo de Confianza:
""")
print(f"""Basado en nuestra muestra, tenemos un {confidence_level*100}% de confianza en que el verdadero valor de la media poblacional (μ) se encuentra entre {lower_bound_ci:.4f} y {upper_bound_ci:.4f}.

Esto significa que si repitiéramos este estudio muchas veces, el 95% de los intervalos de confianza construidos de esta manera contendrían la verdadera media poblacional. La probabilidad se refiere al método, no a un intervalo específico. El intervalo nos da un rango de valores plausibles para μ.
""")


# @title 5. Cálculo del Tamaño de Muestra
# @markdown Aprende por qué es importante el tamaño de muestra y cómo calcularlo para diferentes propósitos estadísticos.

print("""## Cálculo del Tamaño de Muestra

Determinar el tamaño de muestra apropiado es crucial para la validez y eficiencia de un estudio.

### Importancia del Tamaño de Muestra:

*   **Precisión:** Un 'n' más grande mejora la precisión de las estimaciones.
*   **Poder:** Un 'n' adecuado aumenta la probabilidad de detectar efectos reales en pruebas de hipótesis (poder estadístico).
*   **Generalización:** Permite generalizar los hallazgos de la muestra a la población.
*   **Eficiencia:** Optimiza el uso de recursos.

### Factores que Influyen en el Tamaño de Muestra:

*   **Margen de Error Deseado (E):** Tolerancia de diferencia entre la estimación muestral y el parámetro poblacional. Menor E requiere mayor 'n'.
*   **Nivel de Confianza Deseado:** Mayor confianza (ej. 99% vs 95%) requiere mayor 'n'.
*   **Variabilidad Poblacional (σ o s):** Mayor variabilidad requiere mayor 'n'.
*   **Poder Deseado (1 - β):** Mayor poder (ej. 90% vs 80%) requiere mayor 'n' (para pruebas de hipótesis).
*   **Nivel de Significancia (α):** Alpha más pequeño (ej. 0.01 vs 0.05) puede requerir mayor 'n' (para mantener el poder).
*   **Tamaño del Efecto (d):** Magnitud de la diferencia a detectar. Un efecto más pequeño requiere mayor 'n'.

### Cálculo para Estimar una Media Poblacional

Objetivo: Estimar la media poblacional con un cierto nivel de confianza y un margen de error máximo aceptable.

Fórmula: n = (Z * σ / E)²
""")

# @title Calcular Tamaño de Muestra para Estimación de Media
# @markdown Calcula el tamaño de muestra requerido para estimar la media poblacional bajo diferentes escenarios.

# Fórmula para tamaño de muestra para estimar media: n = (Z * σ / E)^2
print("Fórmula utilizada: n = (Z * σ / E)^2")
print("Donde:")
print("  n = Tamaño de muestra requerido")
print("  Z = Puntuación Z para el nivel de confianza deseado")
print("  σ = Desviación estándar de la población (o una estimación)")
print("  E = Margen de error máximo aceptable\n")


# Ejemplo de Cálculo 1: Estimar media con 95% confianza, margen de error 5, desv. estándar 10
desired_confidence_level = 0.95
desired_margin_of_error = 5
estimated_sigma = 10 # Necesitamos una estimación de la desviación estándar poblacional

# Encontrar la puntuación Z para el nivel de confianza deseado (bilateral)
alpha_ss = 1 - desired_confidence_level
z_score_estimate = norm.ppf(1 - alpha_ss / 2) # Z para el percentil (1 - alpha/2)

# Calcular el tamaño de muestra requerido
# Usamos math.ceil para redondear hacia arriba, ya que el tamaño de muestra debe ser un entero.
n_estimate_mean = math.ceil((z_score_estimate * estimated_sigma / desired_margin_of_error) ** 2)

print(f"Ejemplo 1: Estimar la media poblacional")
print(f"  Nivel de confianza deseado: {desired_confidence_level*100}%")
print(f"  Margen de error deseado (E): {desired_margin_of_error}")
print(f"  Desviación estándar estimada (σ): {estimated_sigma}")
print(f"  Puntuación Z para {desired_confidence_level*100}%: {z_score_estimate:.4f}")
print(f"  Tamaño de muestra requerido: ({z_score_estimate:.4f} * {estimated_sigma} / {desired_margin_of_error})^2 = {n_estimate_mean}\n")


# Ejemplo de Cálculo 2: Parámetros diferentes
desired_confidence_level_ex2 = 0.99
desired_margin_of_error_ex2 = 2
estimated_sigma_ex2 = 15

alpha_ss_ex2 = 1 - desired_confidence_level_ex2
z_score_estimate_ex2 = norm.ppf(1 - alpha_ss_ex2 / 2)
n_estimate_mean_ex2 = math.ceil((z_score_estimate_ex2 * estimated_sigma_ex2 / desired_margin_of_error_ex2) ** 2)

print(f"Ejemplo 2: Estimar la media poblacional con parámetros diferentes")
print(f"  Nivel de confianza deseado: {desired_confidence_level_ex2*100}%")
print(f"  Margen de error deseado (E): {desired_margin_of_error_ex2}")
print(f"  Desviación estándar estimada (σ): {estimated_sigma_ex2}")
print(f"  Puntuación Z para {desired_confidence_level_ex2*100}%: {z_score_estimate_ex2:.4f}")
print(f"  Tamaño de muestra requerido: ({z_score_estimate_ex2:.4f} * {estimated_sigma_ex2} / {desired_margin_of_error_ex2})^2 = {n_estimate_mean_ex2}\n")

print("""**Observaciones:**
- Un margen de error más pequeño (Ejemplo 2 vs Ejemplo 1) requiere un tamaño de muestra significativamente mayor.
- Un nivel de confianza más alto (Ejemplo 2 vs Ejemplo 1) requiere un tamaño de muestra mayor.
- Una mayor variabilidad (σ) en la población también requiere un tamaño de muestra mayor.
""")

# @title Calcular Tamaño de Muestra para Prueba de Hipótesis (Test Z Unilateral)
# @markdown Calcula el tamaño de muestra requerido para realizar una prueba de hipótesis con un poder y nivel de significancia deseados.

print("""### Cálculo para una Prueba de Hipótesis (Test Z Unilateral para la Media)

Objetivo: Determinar el tamaño de muestra necesario para detectar una diferencia mínima (tamaño del efecto) con un poder y nivel de significancia especificados.

Fórmula utilizada (Test Unilateral, σ conocida): n = ((Z_α + Z_β) * σ / δ)²
""")

# Fórmula para tamaño de muestra (test unilateral, sigma conocida):
# n = ((Z_alpha + Z_beta) * σ / δ)^2
print("Fórmula utilizada (Test Unilateral): n = ((Z_α + Z_β) * σ / δ)^2")
print("Donde:")
print("  n = Tamaño de muestra requerido")
print("  Z_α = Puntuación Z para el nivel de significancia (α)")
print("  Z_β = Puntuación Z para el poder deseado (1 - β). Beta (β) es la probabilidad de un error Tipo II.")
print("  σ = Desviación estándar de la población (o una estimación previa)")
print("  δ = Diferencia mínima que se desea detectar (Tamaño del efecto en unidades originales)\n")

# Nota: Para un test bilateral, la fórmula usa Z_(alpha/2).
# La fórmula también se puede expresar en términos de tamaño del efecto estandarizado (Cohen's d = δ / σ):
# n = ((Z_alpha + Z_beta) / d)^2
print("Alternativamente, usando el tamaño del efecto estandarizado (d = δ / σ): n = ((Z_α + Z_β) / d)^2\n")


# Ejemplo de Cálculo 1: Test Z Unilateral
# Queremos detectar una diferencia de 3 unidades de la media nula (e.g., H0: mu=50, queremos detectar si mu=53 o más)
# Nivel de significancia (alpha): 0.05 (test unilateral)
# Poder deseado: 0.80 (beta = 0.20)
# Desviación estándar poblacional estimada (sigma): 10
# Diferencia mínima a detectar (delta): 3

significance_level = 0.05
desired_power = 0.80
estimated_sigma = 10
minimum_detectable_difference = 3 # delta

# Encontrar la puntuación Z para alpha (para un test unilateral, usamos 1-alpha)
z_alpha = norm.ppf(1 - significance_level)

# Encontrar la puntuación Z para beta (para calcular el poder, usamos 1-beta)
beta = 1 - desired_power
z_beta = norm.ppf(1 - beta) # Este es el Z-score correspondiente al percentil del poder deseado

# Calcular el tamaño de muestra requerido
n_hypothesis_test = math.ceil(((z_alpha + z_beta) * estimated_sigma / minimum_detectable_difference) ** 2)

print(f"Ejemplo 1: Prueba de Hipótesis (Test Z Unilateral)")
print(f"  Nivel de significancia (α): {significance_level}")
print(f"  Poder deseado (1 - β): {desired_power}")
print(f"  Desviación estándar estimada (σ): {estimated_sigma}")
print(f"  Diferencia mínima a detectar (δ): {minimum_detectable_difference}")
print(f"  Puntuación Z para α ({significance_level}): {z_alpha:.4f}")
print(f"  Puntuación Z para 1-β ({desired_power}): {z_beta:.4f}")
print(f"  Tamaño de muestra requerido: (({z_alpha:.4f} + {z_beta:.4f}) * {estimated_sigma} / {minimum_detectable_difference})^2 = {n_hypothesis_test}\n")


# Otro Ejemplo: Parámetros diferentes
# Nivel de significancia (alpha): 0.01 (unilateral)
# Poder deseado: 0.95
# Desviación estándar poblacional estimada (sigma): 15
# Diferencia mínima a detectar (delta): 2

significance_level_ex2 = 0.01
desired_power_ex2 = 0.95
estimated_sigma_ex2 = 15
minimum_detectable_difference_ex2 = 2

z_alpha_ex2 = norm.ppf(1 - significance_level_ex2)
beta_ex2 = 1 - desired_power_ex2
z_beta_ex2 = norm.ppf(1 - beta_ex2)
n_hypothesis_test_ex2 = math.ceil(((z_alpha_ex2 + z_beta_ex2) * estimated_sigma_ex2 / minimum_detectable_difference_ex2) ** 2)


print(f"Otro Ejemplo: Prueba de Hipótesis (Test Z Unilateral) con parámetros diferentes")
print(f"  Nivel de significancia (α): {significance_level_ex2}")
print(f"  Poder deseado (1 - β): {desired_power_ex2}")
print(f"  Desviación estándar estimada (σ): {estimated_sigma_ex2}")
print(f"  Diferencia mínima a detectar (δ): {minimum_detectable_difference_ex2}")
print(f"  Puntuación Z para α ({significance_level_ex2}): {z_alpha_ex2:.4f}")
print(f"  Puntuación Z para 1-β ({desired_power_ex2}): {z_beta_ex2:.4f}")
print(f"  Tamaño de muestra requerido: (({z_alpha_ex2:.4f} + {z_beta_ex2:.4f}) * {estimated_sigma_ex2} / {minimum_detectable_difference_ex2})^2 = {n_hypothesis_test_ex2}\n")

print("""**Observaciones:**
- Un poder deseado más alto (Ejemplo 2 vs Ejemplo 1) requiere un tamaño de muestra mayor.
- Un nivel de significancia más estricto (alpha más pequeño, Ejemplo 2 vs Ejemplo 1) requiere un tamaño de muestra mayor.
- La detección de una diferencia más pequeña (delta más pequeño, Ejemplo 2 vs Ejemplo 1) requiere un tamaño de muestra significativamente mayor.
- Una mayor variabilidad (σ) también requiere un tamaño de muestra mayor.
""")

# @title Cálculo del Tamaño de Muestra usando statsmodels (Opcional)
# @markdown statsmodels es una librería más completa para análisis de poder y cálculo de tamaño de muestra.

if _STATMODELS_AVAILABLE:
    print("""### Cálculo del Tamaño de Muestra usando statsmodels (Alternativa)

La librería `statsmodels` ofrece funciones más avanzadas para el análisis de poder y el cálculo del tamaño de muestra para una variedad de pruebas.

Usaremos `statsmodels.stats.power.NormalIndPower` para calcular el tamaño de muestra para un test Z unilateral para la media. Requiere especificar el **tamaño del efecto estandarizado (Cohen's d = δ / σ)**.
""")

    # Ejemplo usando statsmodels para test Z unilateral para la media
    # Para el primer ejemplo: delta=3, sigma=10, entonces Cohen's d = 3/10 = 0.3 (tamaño del efecto medio)

    effect_size_statmodels = minimum_detectable_difference / estimated_sigma
    alpha_statmodels = significance_level
    power_statmodels = desired_power
    ratio = 0 # ratio de tamaños de muestra, 0 para test de una muestra

    # Crear un objeto de análisis de poder para distribución normal (test Z)
    power_analysis = NormalIndPower()

    # Calcular tamaño de muestra (para un test de una muestra, es el total)
    # 'solve_for' = 'nobs1' calcula el tamaño de muestra para el primer grupo (nuestra única muestra)
    # 'alternative' especifica el tipo de test ('two-sided', 'larger' o 'smaller')
    n_statmodels = power_analysis.solve_power(
        effect_size=effect_size_statmodels,
        alpha=alpha_statmodels,
        power=power_statmodels,
        ratio=ratio, # Usar ratio=0 para test de una muestra
        alternative='larger' # Opciones válidas: 'two-sided', 'larger', 'smaller'. Usamos 'larger' asumiendo el ejemplo unilateral.
    )

    print(f"Ejemplo usando statsmodels (Test Z Unilateral para la Media, alternativa='larger'):")
    print(f"  Tamaño del efecto (Cohen's d = δ/σ): {effect_size_statmodels:.4f}")
    print(f"  Nivel de significancia (α): {alpha_statmodels}")
    print(f"  Poder deseado (1 - β): {power_statmodels}")
    print(f"  Tamaño de muestra requerido: {math.ceil(n_statmodels)}\n")


    # Ejemplo usando statsmodels para el segundo escenario
    effect_size_statmodels_ex2 = minimum_detectable_difference_ex2 / estimated_sigma_ex2
    alpha_statmodels_ex2 = significance_level_ex2
    power_statmodels_ex2 = desired_power_ex2

    n_statmodels_ex2 = power_analysis.solve_power(
        effect_size=effect_size_statmodels_ex2,
        alpha=alpha_statmodels_ex2,
        power=power_statmodels_ex2,
        ratio=ratio,
        alternative='larger' # Usamos 'larger' asumiendo el ejemplo unilateral.
    )
    print(f"Otro Ejemplo usando statsmodels (Test Z Unilateral, alternativa='larger') con parámetros diferentes:")
    print(f"  Tamaño del efecto (Cohen's d = δ/σ): {effect_size_statmodels_ex2:.4f}")
    print(f"  Nivel de significancia (α): {alpha_statmodels_ex2}")
    print(f"  Poder deseado (1 - β): {power_statmodels_ex2}")
    print(f"  Tamaño de muestra requerido: {math.ceil(n_statmodels_ex2)}\n")

else:
    print("La librería statsmodels no está disponible. Por favor, instálala para ejecutar esta sección.")


print("""### Explicación de los Parámetros para Tamaño de Muestra (Pruebas de Hipótesis):

*   **Nivel de Significancia (α):** Probabilidad de Error Tipo I. Un α más pequeño requiere mayor 'n' para mantener el poder.
*   **Poder (1 - β):** Probabilidad de Error Tipo II (β). Mayor poder requiere mayor 'n'.
*   **Desviación Estándar (σ):** Variabilidad. Mayor σ requiere mayor 'n'.
*   **Diferencia Mínima a Detectar (δ) / Tamaño del Efecto (d):** Magnitud del efecto a detectar. Un efecto más pequeño requiere mayor 'n'.
*   **Z_α y Z_β:** Valores Z de la distribución normal estándar asociados a alpha y beta.

**Consideraciones adicionales:**
- La elección de test unilateral vs bilateral afecta el valor Z_α.
- Si σ es desconocida, se usa una estimación previa.
- `statsmodels` es útil para casos más complejos y diferentes tipos de pruebas.
""")
```
